# Lista de Exercícios 1



Nos exercícios queremos identificar o efeito causal da Educação sobre o Salário. Para isso, vamos usar DAGs (Directed Acyclic Graphs) para ilustrar as relações entre as variáveis.

Carregar pacotes necessários

```python
import graphviz as gr
```

***Exercício 1***

```python
dag1 = gr.Digraph(format='png')
dag1.attr(size='5', rankdir='LR')
dag1.edge('Habilidade', 'Educacao')
dag1.edge('Habilidade', 'Salario')
dag1.edge('Educacao', 'Salario')
dag1
```

**Perguntas:**
* Quais são os caminhos de back-door entre Educação e Salário?
* Quais variáveis precisam ser controladas para identificar o efeito causal de Educação sobre Salário?
* É possível aplicar o critério de front-door neste DAG?

***Exercício 2***

**Perguntas:**

* Se controlarmos por Experiência, o que acontece com a estimativa do efeito total de Educação sobre Salário?
* Qual seria a melhor estratégia de identificação neste caso?

Considere que a variável "Mercado" representa condições econômicas locais (ex.: tamanho do mercado de trabalho, demanda por mão de obra, atividade econômica regional).
