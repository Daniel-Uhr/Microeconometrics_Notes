## 1. Motivação

Em muitos cursos, o aluno aprende regressão, erro-padrão e intervalo de confiança como blocos separados. Isso costuma funcionar operacionalmente, mas produz uma sensação de que a inferência aparece “do nada”.

A proposta desta aula é diferente. Vamos seguir a seguinte trilha:

$$
\text{geometria} \,\longrightarrow\, \text{momento} \,\longrightarrow\, \text{funcional} \,\longrightarrow\, \text{IF} \,\longrightarrow\, \text{inferência} \,\longrightarrow\, \text{causalidade moderna}.
$$

A intuição é simples:

- primeiro entendemos **o problema geométrico**;
- depois mostramos que ele pode ser escrito como **uma condição de momento**;
- em seguida passamos a ver o parâmetro como **uma função da distribuição populacional**;
- então perguntamos: **o que acontece com o parâmetro quando a distribuição muda um pouquinho?**;
- a resposta leva à **função de influência**;
- e a função de influência nos entrega a forma moderna de pensar **erro-padrão, intervalo de confiança, eficiência e escores ortogonais**.

Essa mudança de linguagem é exatamente a ponte para boa parte da econometria causal contemporânea.



## 2. Fundamentos matemáticos mínimos

### 2.1 Conjuntos numéricos

Os conjuntos numéricos mais importantes para a econometria são:

- $\mathbb{N}$: naturais;  
- $\mathbb{Z}$: inteiros;  
- $\mathbb{Q}$: racionais;  
- $\mathbb{R}$: reais;  
- $\mathbb{C}$: complexos.

Para a econometria, o conjunto central é $\mathbb{R}$, porque variáveis, parâmetros, vetores e matrizes costumam ser modelados em espaços reais.

### 2.2 Espaço vetorial

Um **espaço vetorial** é um conjunto de objetos que podem ser somados entre si e multiplicados por escalares, respeitando propriedades algébricas usuais.

O exemplo básico é $\mathbb{R}^k$:

$$
x=(x_1,x_2,\dots,x_k)'.
$$

Casos particulares:

- $\mathbb{R}$: reta real;  
- $\mathbb{R}^2$: plano;  
- $\mathbb{R}^3$: espaço tridimensional;  
- $\mathbb{R}^k$: generalização para dimensão arbitrária.

A ideia importante é esta: mesmo quando não conseguimos “desenhar” $\mathbb{R}^{50}$ ou $\mathbb{R}^{500}$, a estrutura matemática continua perfeitamente definida. Isso é crucial para regressão, GMM, ML e inferência semiparamétrica.



## 3. Espaços vetoriais, produto interno, norma e distância

### 3.1 Produto interno

Em $\mathbb{R}^k$, o produto interno euclidiano entre $x,y\in\mathbb{R}^k$ é

$$
\langle x,y\rangle = x'y = \sum_{j=1}^k x_j y_j.
$$

Interpretação geométrica:

- mede o grau de alinhamento linear entre dois vetores;
- é positivo quando apontam “mais ou menos” na mesma direção;
- é negativo quando apontam em direções opostas;
- é zero quando são ortogonais.

**Cuidado conceitual:** ortogonalidade geométrica **não** é independência estatística. São ideias diferentes.

Uma conexão útil é com a covariância amostral. Se $X$ e $Y$ forem vetores de observações e $\bar X$, $\bar Y$ suas médias, então

$$
\operatorname{cov}(X,Y)=\frac{1}{n}\langle X-\bar X\mathbf 1,\,Y-\bar Y\mathbf 1\rangle.
$$

Ou seja, a covariância pode ser lida como um produto interno entre vetores centrados.

### 3.2 Norma

A norma euclidiana de $x\in\mathbb{R}^k$ é

$$
\|x\| = \sqrt{\langle x,x\rangle} = \sqrt{\sum_{j=1}^k x_j^2}.
$$

Ela mede o comprimento do vetor.

Exemplo:

$$
x=(3,4)' \quad\Rightarrow\quad \|x\|=5.
$$

### 3.3 Distância

A norma induz uma métrica:

$$
d(x,y)=\|x-y\|.
$$

Portanto, distância é apenas o comprimento do vetor diferença. Essa ideia será central quando interpretarmos o OLS como o ponto do espaço coluna de $X$ mais próximo de $y$.



## 4. Projeção ortogonal e interpretação geométrica do OLS

Considere um vetor não nulo $x\in\mathbb{R}^k$. O subespaço gerado por $x$ é a reta

$$
\mathcal{L}(x)=\{ax:a\in\mathbb{R}\}.
$$

Projetar $y$ sobre essa reta significa encontrar o ponto de $\mathcal{L}(x)$ mais próximo de $y$:

$$
\min_{a\in\mathbb{R}}\|y-ax\|^2.
$$

O quadrado aparece apenas para facilitar a derivação; ele não altera o minimizador.

### 4.1 Derivação

Expanda a função objetivo:

$$
\|y-ax\|^2
=\langle y-ax, y-ax\rangle
=\langle y,y\rangle - 2a\langle y,x\rangle + a^2\langle x,x\rangle.
$$

Derivando em relação a $a$ e igualando a zero:

$$
-2\langle y,x\rangle + 2a\langle x,x\rangle = 0.
$$

Logo,

$$
a^*=\frac{\langle y,x\rangle}{\langle x,x\rangle} = \frac{x'y}{x'x}.
$$

Portanto,

$$
\boxed{\operatorname{proj}_x(y)=\frac{x'y}{x'x}x.}
$$

### 4.2 O resíduo é ortogonal

Defina

$$
\hat y = \operatorname{proj}_x(y),
\qquad
r=y-\hat y.
$$

Então

$$
\langle x,r\rangle = 0.
$$

Essa é a ideia-chave: o erro da projeção é perpendicular ao subespaço. Mais adiante, isso reaparecerá como a condição normal do OLS.

### 4.3 Exemplo simples

Se

$$
y=(3,4)', \qquad x=(1,0)',
$$

então

$$
x'y = 3\cdot 1 + 4\cdot 0 = 3,
\qquad
x'x = 1.
$$

Logo,

$$
\operatorname{proj}_x(y)=\frac{3}{1}(1,0)'=(3,0)'.
$$

O resíduo é

$$
r=(3,4)'-(3,0)'=(0,4)',
$$

e de fato

$$
\langle x,r\rangle = (1,0)\cdot(0,4)=0.
$$

**Intuição:** estamos decompondo $y$ em duas partes: uma parte explicada pela direção de $x$ e outra parte que sobra e não pode ser explicada por essa direção.


Vejamos uma representação gráfica dessa projeção em Python:

```python

import matplotlib.pyplot as plt
import numpy as np

# Vetores
y = np.array([3, 4])
x = np.array([1, 0])

# Projeção de y em x
proj_y_em_x = (np.dot(y, x) / np.dot(x, x)) * x

# Resíduo
u = y - proj_y_em_x

plt.figure(figsize=(7, 5))
plt.axhline(0, linestyle=':', color='blue')
plt.axvline(0, linestyle=':', color='blue')

plt.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(
    0, 0,
    proj_y_em_x[0], proj_y_em_x[1],
    angles='xy', scale_units='xy', scale=1,
    linestyle='dashed', color='red'
)

plt.plot([y[0], y[0]], [proj_y_em_x[1], y[1]], linestyle='--')
plt.text(y[0] + 0.1, y[1], 'y=(3,4)')
plt.text(x[0] + 0.1, x[1] + 0.1, 'x=(1,0)')
plt.text(proj_y_em_x[0] + 0.1, proj_y_em_x[1] - 0.3, 'proj_x(y)=(3,0)')
plt.text(y[0] + 0.1, (proj_y_em_x[1] + y[1]) / 2, 'u=(0,4)')

plt.xlim(-0.5, 4.5)
plt.ylim(-0.5, 4.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Projeção ortogonal de y em x')
plt.show()
```


### Como ler a figura

- a seta que vai da origem até $(3,4)$ representa o vetor observado $y$;
- a reta horizontal é o subespaço gerado por $x=(1,0)'$;
- o ponto $(3,0)$ é a projeção ortogonal de $y$ sobre esse subespaço;
- o segmento vertical é o resíduo $u=y-\operatorname{proj}_x(y)$;
- como esse segmento faz $90^\circ$ com o subespaço, temos $\langle u,x\rangle=0$.

Essa figura resume visualmente a ideia do MQO: o estimador escolhe a projeção que minimiza a distância quadrática entre o vetor observado e o espaço gerado pelos regressores.



## 5. Momento: da mecânica à estatística

A palavra “momento” aparece em contextos diferentes:

- em física, pode significar quantidade de movimento;
- em mecânica, pode significar momento de uma força em relação a um pivô;
- em estatística e econometria, passa a significar médias, médias de potências e, mais geralmente, **condições de momento**.

A ponte intuitiva útil é a da gangorra.

### 5.1 Momento em torno de um pivô

Suponha massas $m_i$ colocadas em posições $y_i$ sobre uma reta, e um pivô no ponto $c$. O momento de primeira ordem em torno de $c$ é

$$
M_1(c)=\sum_{i=1}^n m_i(y_i-c).
$$

- se $y_i>c$, a contribuição é positiva;
- se $y_i<c$, a contribuição é negativa.

A condição de equilíbrio é

$$
M_1(c)=0.
$$

Resolvendo para $c$:

$$
\sum_{i=1}^n m_i(y_i-c)=0
\quad\Longrightarrow\quad
c=\frac{\sum_{i=1}^n m_i y_i}{\sum_{i=1}^n m_i}.
$$

Ou seja, o pivô que equilibra a régua é o centro de massa.

**Intuição:** essa já é, na prática, uma condição de momento. O objeto central é uma média ponderada de “desvios em relação ao centro”.



## 6. Esperança, centro de massa e momentos estatísticos

### 6.1 De massas para pesos

Se normalizarmos as massas,

$$
w_i=\frac{m_i}{\sum_{j=1}^n m_j},
\qquad
\sum_{i=1}^n w_i=1,
$$

então o centro de massa vira

$$
\bar y_w = \sum_{i=1}^n w_i y_i,
$$

e a condição de equilíbrio pode ser escrita como

$$
\sum_{i=1}^n w_i(y_i-\bar y_w)=0.
$$

Essa expressão já tem “cara” de estatística.

### 6.2 Centro de massa como esperança

Se $Y$ é discreta e assume valores $y_1,\dots,y_K$ com probabilidades $p_k=P(Y=y_k)$, então

$$
E[Y]=\sum_{k=1}^K p_k y_k.
$$

Logo, a esperança é exatamente o centro de massa da distribuição.

No caso empírico, se cada observação recebe peso $1/n$,

$$
\bar Y = \frac{1}{n}\sum_{i=1}^n Y_i,
$$

isto é, a média amostral é o centro de massa da distribuição empírica.

### 6.3 Momentos estatísticos

O momento bruto de ordem $k$ é

$$
\mu_k' = E[Y^k].
$$

O momento central de ordem $k$ é

$$
\mu_k = E[(Y-E[Y])^k].
$$

Casos importantes:

- $k=1$: momento central igual a zero;  
- $k=2$: variância;  
- $k=3$: assimetria (após normalização);  
- $k=4$: curtose (após normalização).

A mensagem importante é a seguinte: em econometria, “momento” não significa apenas potência de uma variável; ele também significa uma **restrição populacional** que caracteriza o parâmetro de interesse.



## 7. Condições de momento em econometria

Uma condição de momento populacional tem a forma

$$
E_P[\psi(W,\theta_0)]=0,
$$

em que:

- $W$ é o vetor de observáveis;
- $\theta_0$ é o parâmetro verdadeiro;
- $\psi(W,\theta)$ é a função de momento ou equação de estimação;
- $P$ é a distribuição populacional de $W$.

A ideia central é esta:

> o parâmetro verdadeiro é o valor que zera uma restrição populacional.

Isso é muito mais geral do que regressão linear. GMM, IPW, DR, EIF e boa parte da econometria causal moderna podem ser escritos nessa linguagem.

### 7.1 Análogo amostral

Como $P$ é desconhecida, substituímos a esperança pela média amostral:

$$
\frac{1}{n}\sum_{i=1}^n \psi(W_i,\hat\theta)=0.
$$

Essa troca — momento populacional por momento amostral — é o coração de boa parte da econometria.



## 8. OLS como projeção e como equação de momento

Considere o modelo amostral

$$
\mathbf y = \mathbf X\beta + \mathbf u,
$$

com $\mathbf y\in\mathbb{R}^n$, $\mathbf X\in\mathbb{R}^{n\times k}$ e $\beta\in\mathbb{R}^k$.

O OLS resolve

$$
\hat\beta = \arg\min_{\beta\in\mathbb{R}^k}\|\mathbf y-\mathbf X\beta\|^2.
$$

Definindo o resíduo estimado $\hat{\mathbf u}=\mathbf y-\mathbf X\hat\beta$, a condição de primeira ordem é

$$
\mathbf X'(\mathbf y-\mathbf X\hat\beta)=0
\quad\Longleftrightarrow\quad
\mathbf X'\hat{\mathbf u}=0.
$$

### 8.1 Interpretação geométrica

O vetor de resíduos é ortogonal ao espaço coluna de $\mathbf X$. Ou seja, o OLS escolhe a projeção de $\mathbf y$ sobre o espaço gerado pelos regressores.

### 8.2 Interpretação como momento amostral

Escrevendo por observação,

$$
\hat u_i = y_i - x_i'\hat\beta,
$$

obtemos

$$
\frac{1}{n}\sum_{i=1}^n x_i(y_i-x_i'\hat\beta)=0.
$$

Se definirmos

$$
\psi(W_i,\beta)=x_i(y_i-x_i'\beta),
$$

então o OLS resolve

$$
\boxed{\frac{1}{n}\sum_{i=1}^n \psi(W_i,\hat\beta)=0.}
$$

### 8.3 O momento populacional

No modelo populacional,

$$
Y = X'\beta_0 + \varepsilon,
$$

uma condição típica de identificação é

$$
E[\varepsilon\mid X]=0.
$$

Isso implica

$$
E[X\varepsilon]=0
\quad\Longrightarrow\quad
E[X(Y-X'\beta_0)]=0.
$$

Portanto, sob exogeneidade, $\beta_0$ satisfaz um momento populacional.

**Mensagem importante:** o OLS pode ser lido de dois modos ao mesmo tempo:

- como problema geométrico de projeção;
- como solução de uma condição de momento.

Essa dupla leitura é extremamente poderosa.



## 9. Expectativa como funcional da distribuição

Quando escrevemos

$$
E_P[g(W)],
$$

estamos enfatizando que a expectativa depende da distribuição $P$.

Se mudarmos $P$, mudamos a expectativa, mesmo que a função $g$ permaneça igual.

Por isso, podemos ver a expectativa como um **funcional**:

$$
T(P)=\int g(w)\,dP(w).
$$

Um funcional é simplesmente uma aplicação que recebe uma distribuição e devolve um número (ou vetor).

Exemplos:

- média: $T(P)=E_P[Y]$;
- variância: $T(P)=Var_P(Y)$;
- coeficiente de regressão: $T(P)=\beta(P)$.

Essa mudança de linguagem parece abstrata num primeiro momento, mas ela é decisiva. Quando pensamos o parâmetro como funcional da distribuição, passamos a poder perguntar:

> Como esse parâmetro reage a pequenas perturbações em $P$?

Essa é exatamente a pergunta que leva à função de influência.



## 10. Parâmetro como funcional

Em vez de pensar apenas em “um número desconhecido”, vamos pensar no parâmetro como um objeto gerado pela distribuição:

$$
\theta_0 = \Theta(P).
$$

Se o estimador usa a distribuição empírica $\hat P_n$, então

$$
\hat\theta = \Theta(\hat P_n).
$$

Logo,

$$
\hat\theta-\theta_0 = \Theta(\hat P_n)-\Theta(P).
$$

Esse jeito de escrever já prepara a teoria assintótica moderna. Em vez de perguntar apenas “qual é a fórmula do estimador?”, passamos a perguntar “qual é a sensibilidade do funcional $\Theta$ a pequenas mudanças na distribuição?”.



## 11. Contaminação local da distribuição e função de influência

Escolha um ponto $w$ no espaço amostral. Considere a distribuição contaminada

$$
P_\varepsilon = (1-\varepsilon)P + \varepsilon\,\delta_w,
\qquad
0<\varepsilon<1,
$$

em que $\delta_w$ é a medida de Dirac que coloca toda a massa no ponto $w$.

### Intuição

Estamos perguntando:

> O que acontece com o parâmetro se eu der um peso infinitesimal a uma observação específica?

Se o funcional $\Theta$ for diferenciável em uma vizinhança de $P$, definimos a **função de influência** por

$$
IF(w;\Theta,P)=\left.\frac{d}{d\varepsilon}\Theta(P_\varepsilon)\right|_{\varepsilon=0}.
$$

A função de influência mede o efeito marginal de uma contaminação local sobre o parâmetro.

### Leitura prática

- se $IF(w)$ é grande e positivo, aquele tipo de observação empurra o estimador para cima;
- se $IF(w)$ é grande e negativo, empurra para baixo;
- se $IF(w)$ é próximo de zero, tem pouca influência local.

É essa ideia que, depois, alimenta variância assintótica, erro-padrão robusto, eficiência e escores ortogonais.



## 12. Exemplo: função de influência da média

Considere o funcional da média:

$$
\mu(P)=E_P[Y].
$$

Sob a distribuição contaminada,

$$
\mu(P_\varepsilon)=(1-\varepsilon)E_P[Y]+\varepsilon y.
$$

Derivando em $\varepsilon=0$,

$$
IF(y;\mu,P)=y-E_P[Y].
$$

Logo,

$$
\boxed{IF(y;\mu,P)=y-\mu.}
$$

Essa expressão é extremamente intuitiva: a influência local de uma observação sobre a média é apenas o seu desvio em relação à média populacional.



## 13. Exemplo: função de influência do OLS

Considere $W=(Y,X)$, com $X\in\mathbb{R}^p$, e defina o funcional do OLS por

$$
\beta(P)=\arg\min_{b\in\mathbb{R}^p} E_P[(Y-X'b)^2].
$$

A condição de primeira ordem populacional é

$$
E_P[X(Y-X'\beta)]=0.
$$

Se definirmos

$$
A=E_P[XX'],
\qquad
c=E_P[XY],
$$

então, sob invertibilidade de $A$,

$$
\beta(P)=A^{-1}c.
$$

### 13.1 Contaminação local

Considere $w=(y,x)$ e

$$
P_\varepsilon=(1-\varepsilon)P+\varepsilon\delta_{(y,x)}.
$$

Sob $P_\varepsilon$,

$$
A_\varepsilon=(1-\varepsilon)A+\varepsilon xx',
\qquad
c_\varepsilon=(1-\varepsilon)c+\varepsilon xy.
$$

Logo,

$$
\beta_\varepsilon=A_\varepsilon^{-1}c_\varepsilon.
$$

### 13.2 Derivação

Usando a regra da derivada da inversa matricial,

$$
\frac{d}{d\varepsilon}(A_\varepsilon^{-1})
=
-A_\varepsilon^{-1}\left(\frac{dA_\varepsilon}{d\varepsilon}\right)A_\varepsilon^{-1},
$$

obtemos, em $\varepsilon=0$,

$$
\left.\frac{d\beta_\varepsilon}{d\varepsilon}\right|_{\varepsilon=0}
=
A^{-1}x(y-x'\beta).
$$

Portanto,

$$
\boxed{IF((y,x);\beta,P)=A^{-1}x(y-x'\beta).}
$$

### 13.3 Interpretação

O termo $u=y-x'\beta$ é o resíduo populacional. Assim, a IF do OLS combina:

- **resíduo**: quão “surpreendente” é a observação em termos do ajuste populacional;
- **alavancagem direcional** $A^{-1}x$: como essa surpresa é transmitida para o coeficiente.

Em linguagem intuitiva, a observação influencia mais o OLS quando tem grande erro e/ou está em uma direção do espaço de regressoras que altera muito a geometria do problema.



## 14. Relação entre equação de estimação, Jacobiano e função de influência

Para muitos M-estimadores, o parâmetro é definido por

$$
E[\psi(W,\theta_0)]=0.
$$

Sob regularidade, a função de influência associada é

$$
\boxed{IF(w;\theta_0,P)= -J^{-1}\psi(w,\theta_0),}
$$

em que

$$
J=E[\partial_\theta \psi(W,\theta_0)]
$$

é o Jacobiano populacional da equação de estimação.

### Intuição

- $\psi(w,\theta_0)$ é a contribuição daquela observação para o momento;
- $J$ diz como o momento reage a mudanças em $\theta$;
- $-J^{-1}$ transforma a contribuição do momento em contribuição sobre o estimador.

No OLS,

$$
\psi(W,\beta)=X(Y-X'\beta),
\qquad
J=-E[XX']=-A,
$$

logo

$$
IF((Y,X);\beta,P)=A^{-1}X(Y-X'\beta),
$$

como obtido acima.

Essa fórmula é muito importante porque reaparece em praticamente toda a econometria moderna.


## 15. Eficiência semiparamétrica e EIF: uma primeira visão

Até aqui, vimos que a **função de influência (IF)** descreve a sensibilidade local de um parâmetro a pequenas perturbações da distribuição.

Neste ponto, surge uma pergunta natural:

> Se existem várias representações válidas dessa sensibilidade local, existe alguma que seja melhor do que as outras?

A resposta é sim. Essa representação especial é a **função de influência eficiente**, ou **EIF** (*efficient influence function*).

A ideia intuitiva é simples. A IF já nos diz como o parâmetro reage localmente a mudanças na distribuição. A EIF acrescenta um critério de qualidade: entre todas as representações válidas, ela é a que leva à **menor variância assintótica possível** dentro do modelo considerado.

### 15.1 A ideia em palavras

Podemos pensar assim:

- a **IF** responde: *como o parâmetro muda quando a distribuição é ligeiramente perturbada?*
- a **EIF** responde: *qual é a forma mais informativa e mais precisa de representar essa sensibilidade local?*

Portanto, a EIF é uma IF com uma propriedade adicional de otimalidade.

### 15.2 Uma formulação formal enxuta

Se $T(P)$ é o parâmetro de interesse e $\mathcal{T}(P)$ denota o **espaço tangente** do modelo em $P$, então a EIF é uma função $\phi^{\mathrm{eff}}(W)$ tal que, para toda direção admissível $s(W)\in\mathcal{T}(P)$,

$$
\dot T_P[s] = E\big[\phi^{\mathrm{eff}}(W)s(W)\big],
$$

com

$$
E\big[\phi^{\mathrm{eff}}(W)\big]=0,
$$

e com variância mínima entre todas as funções que representam corretamente essa mesma derivada direcional.

Aqui, vale uma leitura intuitiva:

- o **espaço tangente** reúne as direções locais admissíveis de variação da distribuição;
- a derivada $\dot T_P[s]$ mede como o parâmetro reage nessas direções;
- a EIF é a função que representa essa reação da maneira mais eficiente.

Nesta aula, não precisamos desenvolver toda a teoria do espaço tangente em profundidade. O importante é guardar a intuição correta: a EIF é a versão "ótima" da função de influência no modelo considerado.

### 15.3 O que significa “eficiente”?

Eficiência, aqui, significa **menor variância assintótica regular possível**.

Se um estimador regular e assintoticamente linear admite expansão

$$
\sqrt{n}(\hat\theta-\theta_0)
=
\frac{1}{\sqrt{n}}\sum_{i=1}^n \phi(W_i) + o_p(1),
$$

então a variância assintótica desse estimador é determinada por $\mathrm{Var}(\phi(W))$.

Quando $\phi=\phi^{\mathrm{eff}}$, obtemos o menor valor possível dessa variância dentro da classe regular. Por isso,

$$
\mathrm{Var}\big(\phi^{\mathrm{eff}}(W)\big)
$$

define o chamado **limite de eficiência semiparamétrica**.

### 15.4 Por que isso importa?

Porque a EIF nos diz qual é, localmente, a melhor precisão assintótica que podemos esperar de um estimador regular.

Em termos práticos:

- toda IF válida ajuda a descrever a expansão assintótica;
- a EIF faz isso da maneira mais precisa possível;
- por isso, estimadores baseados na EIF, ou em escores equivalentes a ela, tendem a ter papel central em inferência moderna.

Essa é a razão pela qual a EIF aparece, mais adiante, em temas como:

- estimadores duplamente robustos;
- AIPW;
- TMLE;
- DML com escores ortogonais;
- DiD duplamente robusto.

### 15.5 O ponto pedagógico mais importante

Nesta aula, o objetivo não é esgotar a teoria semiparamétrica. O objetivo é fixar a ideia correta:

> depois de entender função de influência, o passo seguinte natural é perguntar se existe uma versão "ótima" dessa sensibilidade local.

A resposta é a EIF.

Ela será retomada em aulas posteriores, quando tratarmos com mais detalhe de:

- derivada *pathwise*;
- espaço tangente;
- gradiente canônico;
- escores ortogonais;
- eficiência semiparamétrica em modelos causais.

## 16. Linearização e expansão de von Mises

Suponha que $W_1,\dots,W_n$ são i.i.d. de $P$ e que

$$
\hat\theta=\Theta(\hat P_n),
\qquad
\theta_0=\Theta(P).
$$

Se $\Theta$ for suficientemente regular, então existe uma expansão do tipo

$$
\Theta(\hat P_n)-\Theta(P)
=
\int IF(w;\Theta,P)\,d(\hat P_n-P)(w)+o_p(n^{-1/2}).
$$

Como

$$
\hat P_n=\frac{1}{n}\sum_{i=1}^n \delta_{W_i},
$$

segue que

$$
\Theta(\hat P_n)-\Theta(P)
=
\frac{1}{n}\sum_{i=1}^n IF(W_i;\Theta,P)
-
E_P[IF(W;\Theta,P)]
+
o_p(n^{-1/2}).
$$

Tomando a versão centrada da IF, com $E_P[IF(W;\Theta,P)]=0$, obtemos

$$
\boxed{
\hat\theta-\theta_0
=
\frac{1}{n}\sum_{i=1}^n IF(W_i;\Theta,P)+o_p(n^{-1/2}).
}
$$

Essa é a representação assintótica fundamental. Ela diz que, na primeira ordem, o erro do estimador se comporta como uma média amostral.

**Intuição:** depois da linearização, a inferência volta a ser, essencialmente, inferência sobre médias.


## 17. Leitura alternativa via Taylor em M-estimadores

Muitos estimadores satisfazem

$$
0=\frac{1}{n}\sum_{i=1}^n \psi(W_i,\hat\theta)=:g_n(\hat\theta),
$$

com momento populacional

$$
g(\theta)=E[\psi(W,\theta)].
$$

Fazendo uma expansão de Taylor de primeira ordem em torno de $\theta_0$,

$$
0 \approx g_n(\theta_0)+\partial_\theta g_n(\theta_0)(\hat\theta-\theta_0).
$$

Isolando $\hat\theta-\theta_0$,

$$
\hat\theta-\theta_0
\approx
-\big(\partial_\theta g_n(\theta_0)\big)^{-1}g_n(\theta_0).
$$

Sob LLN,

$$
\partial_\theta g_n(\theta_0)\to J=E[\partial_\theta\psi(W,\theta_0)].
$$

Logo,

$$
\hat\theta-\theta_0
=
\frac{1}{n}\sum_{i=1}^n\big(-J^{-1}\psi(W_i,\theta_0)\big)+o_p(n^{-1/2}).
$$

Portanto,

$$
IF(w;\theta_0,P)=-J^{-1}\psi(w,\theta_0).
$$

Essa derivação é menos abstrata do que a expansão de von Mises e costuma ser muito útil didaticamente. Ela mostra, com clareza, de onde sai a IF em muitos estimadores usuais.



## 18. Normalidade assintótica, variância assintótica e taxa $n^{-1/2}$

Da representação linear,

$$
\hat\theta-\theta_0
=
\frac{1}{n}\sum_{i=1}^n IF_i + o_p(n^{-1/2}),
\qquad IF_i:=IF(W_i;\Theta,P),
$$

segue que

$$
\sqrt{n}(\hat\theta-\theta_0)
=
\frac{1}{\sqrt{n}}\sum_{i=1}^n IF_i + o_p(1).
$$

Se $E[IF_i]=0$ e $Var(IF_i)=\Sigma<\infty$, então, pelo CLT,

$$
\boxed{\sqrt{n}(\hat\theta-\theta_0)\Rightarrow \mathcal N(0,\Sigma).}
$$

### O que isso significa?

- o erro do estimador tem ordem típica $n^{-1/2}$;
- a variância assintótica do estimador é

$$
Var(\hat\theta)\approx \frac{1}{n}\Sigma;
$$

- portanto, quanto maior a amostra, menor a variância do estimador.

### Intuição importante

A normalidade assintótica não aparece “por mágica”. Ela aparece porque, após a linearização, o erro do estimador vira uma média de contribuições individuais, e médias obedecem ao Teorema Central do Limite.


## 19. Construção de erro-padrão e intervalo de confiança

### 19.1 Estimar a função de influência

Como a IF verdadeira depende de objetos populacionais desconhecidos, usamos plug-in:

$$
\widehat{IF}_i = IF(W_i;\hat\theta,\hat P_n).
$$

### 19.2 Estimar a variância

Para $\theta\in\mathbb{R}^k$,

$$
\boxed{
\widehat{Var}(\hat\theta)
=
\frac{1}{n^2}\sum_{i=1}^n \widehat{IF}_i\widehat{IF}_i'.
}
$$

No caso escalar,

$$
\widehat{Var}(\hat\theta)
=
\frac{1}{n^2}\sum_{i=1}^n \widehat{IF}_i^2,
\qquad
se(\hat\theta)=\sqrt{\widehat{Var}(\hat\theta)}.
$$

### 19.3 Intervalo de confiança Wald

Para o componente $j$ de $\hat\theta$,

$$
\boxed{
IC_{1-\alpha}(\theta_j):
\quad
\hat\theta_j \pm z_{1-\alpha/2}\,se(\hat\theta_j).
}
$$

### 19.4 Caso típico dos M-estimadores

Se

$$
\frac{1}{n}\sum_{i=1}^n \psi(W_i,\hat\theta)=0,
$$

então

$$
\widehat{IF}_i=-\hat J^{-1}\psi(W_i,\hat\theta),
\qquad
\hat J=\frac{1}{n}\sum_{i=1}^n \partial_\theta\psi(W_i,\hat\theta).
$$

### 19.5 Exemplo: OLS robusto à heterocedasticidade

No OLS,

$$
\widehat{IF}_i = \hat A^{-1}X_i\hat u_i,
\qquad
\hat A=\frac{1}{n}\sum_{i=1}^n X_iX_i',
\qquad
\hat u_i = Y_i-X_i'\hat\beta.
$$

Logo,

$$
\widehat{Var}(\hat\beta)
=
\frac{1}{n^2}\sum_{i=1}^n (\hat A^{-1}X_i\hat u_i)(\hat A^{-1}X_i\hat u_i)'.
$$

Essa é, em essência, a lógica por trás do estimador sanduíche robusto.


## 20. Extensão para dados em painel e clusterização

Em muitos problemas aplicados, as observações não são i.i.d. ao nível individual. Em painel, por exemplo, erros costumam ser correlacionados dentro da mesma unidade ao longo do tempo.

Nesses casos, a ideia correta é agregar a contribuição da IF ao nível do cluster.

Se $g=1,\dots,G$ indexa clusters, defina

$$
\hat S_g = \sum_{i\in g}\widehat{IF}_i.
$$

Então a variância cluster-robust pode ser estimada por

$$
\boxed{
\widehat{Var}_{cl}(\hat\theta)
=
\frac{1}{n^2}\sum_{g=1}^G \hat S_g\hat S_g'.
}
$$

**Intuição:** quando há dependência intracluster, a unidade relevante para a variabilidade deixa de ser a observação isolada e passa a ser o cluster.

Essa lógica é muito importante em DiD, event studies, painel e várias aplicações de econometria causal.


## 21. Econometria Causal Moderna: momentos ortogonais, EIF e dupla robustez

Até aqui, desenvolvemos os fundamentos necessários para entender uma ideia central da econometria moderna: muitos estimadores causais podem ser vistos como soluções de condições de momento, e sua inferência pode ser construída a partir de funções de influência.

Na econometria causal contemporânea, porém, surge um desafio adicional. Em geral, o parâmetro de interesse depende de objetos auxiliares desconhecidos — como probabilidades condicionais, regressões condicionais e densidades — que também precisam ser estimados. Esses objetos são chamados de *nuisance functions*.

A grande questão passa a ser:

> Como construir estimadores causais que continuem bem comportados mesmo quando essas funções auxiliares são estimadas de forma flexível?

É exatamente aqui que entram a **função de influência eficiente (EIF)**, a **dupla robustez** e os **momentos ortogonais**.

### 21.1 O problema central da econometria causal moderna

Em muitos problemas causais, o parâmetro de interesse não depende apenas da distribuição marginal observável, mas também de componentes condicionais que precisam ser aprendidos a partir dos dados.

Por exemplo, no caso do efeito médio do tratamento, aparecem naturalmente:

- a probabilidade de receber tratamento dado \(X\), isto é, o **propensity score**;
- as médias condicionais do resultado sob tratamento e sob controle.

Se esses objetos forem estimados com erro, o estimador final também pode ser contaminado. Logo, o problema não é apenas identificar o parâmetro causal, mas também construir um estimador que seja pouco sensível, em primeira ordem, aos erros cometidos na estimação dos componentes auxiliares.

Essa é a motivação para a EIF.

### 21.2 Exemplo central: o efeito médio do tratamento (ATE)

Considere dados observados \(W=(Y,D,X)\), onde:

- \(Y\) é o resultado observado;
- \(D\in\{0,1\}\) indica tratamento;
- \(X\) é o vetor de covariáveis observadas.

Sob as hipóteses usuais de identificação causal — especialmente ignorabilidade condicional e sobreposição — o **efeito médio do tratamento** é

$$
\tau = E[Y(1)-Y(0)].
$$

Sob essas hipóteses, podemos escrever

$$
\tau = E[\mu_1(X)-\mu_0(X)],
$$

onde

$$
\mu_1(X)=E[Y\mid D=1,X], 
\qquad
\mu_0(X)=E[Y\mid D=0,X].
$$

Além disso, definimos o propensity score como

$$
e(X)=P(D=1\mid X).
$$

Portanto, mesmo um parâmetro aparentemente simples já depende de três funções auxiliares:

- \(e(X)\),
- \(\mu_1(X)\),
- \(\mu_0(X)\).

### 21.3 Por que a EIF aparece aqui?

A IF geral nos ensinou como medir a sensibilidade local de um parâmetro. Mas, em problemas causais modernos, queremos algo mais específico:

- uma representação válida da sensibilidade local;
- que organize corretamente a presença das funções auxiliares;
- e que tenha a menor variância assintótica possível dentro do modelo.

Esse objeto é a **função de influência eficiente (EIF)**.

A intuição é a seguinte: a EIF resume, num único objeto, toda a informação relevante sobre o parâmetro causal, separando da melhor forma possível o que é informação sobre o parâmetro daquilo que é apenas variação associada aos *nuisances*.

Por isso, a EIF ocupa papel central em:

- AIPW;
- estimadores duplamente robustos;
- TMLE;
- DML;
- DiD duplamente robusto;
- estimadores modernos de ATT, ATE e CATE.

### 21.4 Derivação heurística da EIF do ATE

Agora vamos mostrar, de forma intuitiva e correta, como surge a EIF do ATE.

Sabemos que

$$
\tau = E[\mu_1(X)-\mu_0(X)].
$$

Se conhecêssemos exatamente \(\mu_1(X)\) e \(\mu_0(X)\), bastaria tomar a média populacional dessa diferença.

Mas isso não utiliza toda a informação observada da amostra. Em particular, podemos decompor o resultado observado em média condicional mais resíduo:

$$
Y = \mu_1(X) + \big(Y-\mu_1(X)\big) \quad \text{quando } D=1,
$$

e

$$
Y = \mu_0(X) + \big(Y-\mu_0(X)\big) \quad \text{quando } D=0.
$$

Esses resíduos têm média zero condicionalmente a \(X\), mas só são observados nos grupos correspondentes. Para transformá-los em correções válidas para toda a população, usamos ponderação pelo propensity score.

A correção do lado tratado é

$$
\frac{D}{e(X)}\big(Y-\mu_1(X)\big),
$$

e a correção do lado controle é

$$
\frac{1-D}{1-e(X)}\big(Y-\mu_0(X)\big).
$$

Logo, somando a diferença média condicional e as correções residuais apropriadas, obtemos a EIF:

$$
\phi^{\mathrm{eff}}(W)
=
\frac{D}{e(X)}\big(Y-\mu_1(X)\big)
-
\frac{1-D}{1-e(X)}\big(Y-\mu_0(X)\big)
+
\mu_1(X)-\mu_0(X)-\tau.
$$

Essa expressão é construída de modo que

$$
E[\phi^{\mathrm{eff}}(W)] = 0.
$$

Esse centramento é essencial: a EIF não é o parâmetro em si, mas a flutuação local do estimador ao redor do parâmetro verdadeiro.

### 21.5 Leitura intuitiva da EIF do ATE

A fórmula acima pode parecer longa à primeira vista, mas sua lógica é muito simples.

O primeiro termo,

$$
\frac{D}{e(X)}\big(Y-\mu_1(X)\big),
$$

captura a parte inesperada do resultado entre os tratados, reponderando pelo inverso da probabilidade de tratamento.

O segundo termo,

$$
-\frac{1-D}{1-e(X)}\big(Y-\mu_0(X)\big),
$$

faz a correção análoga para o grupo controle.

O terceiro bloco,

$$
\mu_1(X)-\mu_0(X)-\tau,
$$

representa a diferença média condicional entre tratamento e controle, centrada no verdadeiro ATE.

Portanto, a EIF combina:

- uma parte de **outcome regression**;
- uma parte de **inverse probability weighting**;
- uma estrutura de correção que torna o escore localmente robusto;
- e a melhor eficiência assintótica regular disponível no modelo.

### 21.6 Da EIF ao estimador

Uma vez conhecida a EIF, ela sugere imediatamente a forma de um estimador. Basta substituir os objetos desconhecidos por estimativas e impor a contrapartida amostral da condição de média zero:

$$
\frac{1}{n}\sum_{i=1}^n \hat{\phi}^{\mathrm{eff}}(W_i)=0.
$$

Equivalentemente, podemos escrever o estimador AIPW/DR do ATE como

$$
\hat{\tau}_{DR}
=
\frac{1}{n}\sum_{i=1}^n
\left[
\hat{\mu}_1(X_i)-\hat{\mu}_0(X_i)
+
\frac{D_i}{\hat e(X_i)}\big(Y_i-\hat{\mu}_1(X_i)\big)
-
\frac{1-D_i}{1-\hat e(X_i)}\big(Y_i-\hat{\mu}_0(X_i)\big)
\right].
$$

Ou seja, o estimador duplamente robusto é simplesmente a média empírica da estrutura sugerida pela EIF.

### 21.7 Por que isso gera dupla robustez?

A propriedade de dupla robustez vem do fato de que a expressão acima continua correta se:

- o propensity score estiver corretamente especificado, mesmo que a regressão de resultado esteja errada; ou
- a regressão de resultado estiver corretamente especificada, mesmo que o propensity score esteja errado.

Isso acontece porque a EIF foi construída de forma a combinar duas representações do parâmetro: uma baseada em reponderação e outra baseada em regressão.

Essa é a intuição fundamental da dupla robustez:

> o estimador tem duas “rotas” de consistência.

### 21.8 Relação com momentos ortogonais

A EIF também ajuda a entender a noção de **momento ortogonal**.

Na econometria moderna, buscamos escores \(\psi(W;\theta,\eta)\) tais que:

$$
E[\psi(W;\theta_0,\eta_0)] = 0,
$$

e cuja derivada em relação às funções auxiliares seja nula em primeira ordem no ponto verdadeiro:

$$
\partial_\eta E[\psi(W;\theta_0,\eta_0)][\eta-\eta_0]=0.
$$

A intuição é clara:

- queremos identificar \(\theta_0\);
- mas também queremos que pequenos erros em \(\eta\) não prejudiquem a estimação de \(\theta_0\) em primeira ordem.

Essa é exatamente a lógica de **Neyman orthogonality**, que sustenta o DML.

Em muitos casos, o escore ortogonal usado em DML coincide com a EIF ou é algebraicamente equivalente a ela.

### 21.9 Variância assintótica da EIF

Uma vez obtida a expansão assintoticamente linear,

$$
\sqrt{n}(\hat{\tau}-\tau)
=
\frac{1}{\sqrt{n}}
\sum_{i=1}^n
\phi^{\mathrm{eff}}(W_i)
+
o_p(1),
$$

o Teorema Central do Limite implica que

$$
\sqrt{n}(\hat{\tau}-\tau)
\overset{d}{\longrightarrow}
N\big(0, V_{\mathrm{eff}}\big),
$$

onde

$$
V_{\mathrm{eff}}
=
\mathrm{Var}\big(\phi^{\mathrm{eff}}(W)\big).
$$

Portanto, a variância assintótica do estimador eficiente é precisamente a variância da EIF.

Equivalentemente,

$$
\mathrm{Avar}(\hat{\tau})
=
\frac{1}{n}\,\mathrm{Var}\big(\phi^{\mathrm{eff}}(W)\big).
$$

Essa é uma conclusão central da teoria:

> a EIF não apenas organiza a forma do estimador; ela também determina diretamente sua melhor variância assintótica regular.

### 21.10 Estimação prática da variância

Na prática, substituímos a EIF verdadeira por sua versão estimada:

$$
\hat{\phi}_i
=
\frac{D_i}{\hat e(X_i)}\big(Y_i-\hat{\mu}_1(X_i)\big)
-
\frac{1-D_i}{1-\hat e(X_i)}\big(Y_i-\hat{\mu}_0(X_i)\big)
+
\hat{\mu}_1(X_i)-\hat{\mu}_0(X_i)-\hat{\tau}.
$$

Então estimamos a variância assintótica por

$$
\widehat{\mathrm{Avar}}(\hat{\tau})
=
\frac{1}{n}
\left(
\frac{1}{n}
\sum_{i=1}^n
\hat{\phi}_i^2
\right),
$$

ou, com correção de centragem amostral,

$$
\widehat{\mathrm{Avar}}(\hat{\tau})
=
\frac{1}{n}
\left(
\frac{1}{n}
\sum_{i=1}^n
(\hat{\phi}_i-\bar{\hat{\phi}})^2
\right).
$$

Assim, o erro-padrão estimado é

$$
\widehat{SE}(\hat{\tau})
=
\sqrt{\widehat{\mathrm{Avar}}(\hat{\tau})}.
$$

Essa etapa fecha perfeitamente o arco da aula: da função de influência à inferência prática.

### 21.11 O que esta seção prepara para as próximas aulas

Esta seção mostra por que a econometria causal moderna pode ser vista como uma continuação natural dos fundamentos desenvolvidos anteriormente.

A sequência lógica é:

1. definimos o parâmetro como funcional;
2. estudamos a sensibilidade local por meio da IF;
3. entendemos a expansão assintoticamente linear;
4. aplicamos isso a um parâmetro causal concreto;
5. obtemos a EIF;
6. derivamos a variância assintótica;
7. e construímos um estimador duplamente robusto e ortogonal.

É exatamente essa arquitetura que reaparece, com adaptações, em:

- ATE;
- ATT;
- DR-DiD;
- DML;
- TMLE;
- event studies modernos;
- modelos com aprendizado de máquina para funções auxiliares.

### 21.12 Síntese final

A principal mensagem é a seguinte:

- a IF mede sensibilidade local;
- a EIF mede essa sensibilidade da forma mais eficiente possível;
- a média amostral da EIF gera estimadores modernos de alta qualidade;
- e a variância da EIF determina a variância assintótica desses estimadores.

Por isso, a EIF não deve ser vista como um detalhe técnico isolado. Ela é um dos objetos centrais que unem:

- teoria de momentos;
- inferência assintótica;
- dupla robustez;
- ortogonalidade;
- e econometria causal moderna.

- o estimador nasce de uma restrição;
- a restrição é escrita como momento;
- o parâmetro pode ser visto como funcional;
