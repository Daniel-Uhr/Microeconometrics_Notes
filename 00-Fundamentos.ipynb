## Fundamentos Matemáticos

### Definições

* **Geometria**: estudo de formas, tamanhos, posições relativas de figuras e das propriedades do espaço.

### Conjuntos Numéricos Importantes

* **$\mathbb{N}$**: conjunto dos números naturais: $\{1, 2, 3, \ldots\}$.
  * Exemplo: contagem de objetos — 1 maçã, 2 laranjas, 3 bananas.

* **$\mathbb{Z}$**: conjunto dos números inteiros: $\{\ldots, -3, -2, -1, 0, 1, 2, 3, \ldots\}$.
  * Exemplo: temperaturas abaixo e acima de zero — -5°C, 0°C, 10°C.

* **$\mathbb{Q}$**: conjunto dos números racionais, que podem ser escritos como fração $\frac{a}{b}$, com $a, b \in \mathbb{Z}$ e $b \neq 0$.
  * Exemplo: $\frac{1}{2}$, $\frac{3}{4}$, $\frac{-5}{2}$.

* **$\mathbb{I}$**: conjunto dos números irracionais, que não podem ser escritos como fração de inteiros.
  * Exemplo: $\sqrt{2}$, $\pi$, $e$.

* **$\mathbb{R}$**: conjunto dos números reais, que inclui os racionais e irracionais.
  * Exemplo: medidas contínuas — altura (1.75 m), peso (70.5 kg).

* **$\mathbb{C}$**: conjunto dos números complexos, da forma $a + bi$, com $a,b \in \mathbb{R}$ e $i^2=-1$.
  * Exemplo: $3+4i$, $-2-5i$.
  * Aplicações em engenharia elétrica e física.

Existem muitos outros conjuntos numéricos, mas **para a econometria**, o conjunto fundamental é o dos **números reais** e os **espaços vetoriais construídos a partir deles**.


## Espaços Vetoriais e Geometria dos Dados

### Espaço Vetorial

Um **espaço vetorial** é um conjunto de objetos, chamados vetores, que podem ser:
- somados entre si;
- multiplicados por escalares;

obedecendo a propriedades como associatividade, comutatividade, existência de elemento neutro e de inverso aditivo.


### O Espaço $\mathbb{R}^k$

O conjunto $\mathbb{R}^k$ não é apenas um conjunto: ele é um **espaço vetorial real de dimensão $k$**.

Cada elemento de $\mathbb{R}^k$ é um vetor:
$$
x = (x_1, x_2, \ldots, x_k)
$$

O espaço admite:
- soma de vetores;
- multiplicação por escalares reais.

Casos particulares:
- $\mathbb{R}^1$: a reta real;
- $\mathbb{R}^2$: o plano cartesiano;
- $\mathbb{R}^3$: o espaço tridimensional;
- $\mathbb{R}^k$, com $k>3$: espaço de dimensão maior, não visualizável, mas perfeitamente bem definido matematicamente.



### Produto Interno

Em $\mathbb{R}^k$, define-se o **produto interno euclidiano** entre dois vetores $x,y \in \mathbb{R}^k$ por:
$$
\langle x,y\rangle = \sum_{j=1}^k x_j y_j
$$

Interpretação geométrica (soma de produtos coordenada a coordenada):
- o produto interno mede o **grau de alinhamento** entre dois vetores;
- é máximo quando os vetores são paralelos;
- é zero quando os vetores são ortogonais (ângulo de 90°). Representa independência direcional.

### Norma (Comprimento do Vetor)

A **norma euclidiana** de um vetor $x \in \mathbb{R}^k$ é definida por:
$$
\|x\| = \sqrt{\sum_{j=1}^k x_j^2}
$$

- a norma transforma um vetor em um número real não negativo;
- geometricamente, ela representa o **comprimento da seta que liga a origem ao ponto $x$**.

Exemplo:
$$
x=(3,4)\in\mathbb{R}^2 \quad \Rightarrow \quad \|x\|=\sqrt{3^2+4^2}=5
$$

### Métrica (Distância entre Pontos)

A norma induz uma **métrica**, isto é, uma noção de distância entre dois pontos $x,y \in \mathbb{R}^k$:
$$
d(x,y)=\|x-y\|
$$

Exemplo simples:
- no espaço $\mathbb{R}$, a métrica usual é:
$$
d(x,y)=|x-y|
$$


### Projeção de Vetores

Vamos entender a projeção de vetores porque ela é fundamental para a compreensão de conceitos como **regressão linear** e **ajuste de modelos**.

O subespaço gerado por ($x\neq 0$) é a reta

$$
\mathcal{L}={a x: a\in\mathbb{R}}.
$$

Projetar ($y$) nessa reta significa achar o ponto da reta ($ax$) **mais próximo** de ($y$) (em norma Euclidiana). Ou seja, resolver:

$$
\min_{a\in\mathbb{R}}\ |y-ax|^2.
$$

Obs: o quadrado não muda o minimizador, só muda a escala da função objetivo, e é mais fácil de derivar.

Chame o minimizador de ($a^{*}$). Então a projeção é
$$
\operatorname{proj}_x(y)=a^*x.
$$

***Derivação por minimização***

Expanda a função objetivo:

$$
|y-ax|^2=\langle y-ax,\ y-ax\rangle
= \langle y,y\rangle -2a\langle y,x\rangle + a^2\langle x,x\rangle.
$$

Isso é uma parábola em ($a$). Derive e iguale a zero:

$$
\frac{d}{da}|y-ax|^2 = -2\langle y,x\rangle +2a\langle x,x\rangle=0.
$$

Logo,

$$
a^*=\frac{\langle y,x\rangle}{\langle x,x\rangle}
=\frac{\langle y,x\rangle}{|x|^2}.
$$

Portanto,

$$
\boxed{\operatorname{proj}_x(y)=\frac{\langle y,x\rangle}{|x|^2} x}
$$

Essa é exatamente a fórmula da projeção de um vetor sobre outro.


***Interpretação geométrica: o “erro” é perpendicular***

Defina

$$
\hat y=\operatorname{proj}_x(y)=a^*x,\qquad r=y-\hat y.
$$

A condição de otimalidade acima é equivalente a:

$$
\langle x,\ r\rangle =0.
$$

Ou seja, o **resíduo** (r) (a parte de (y) que sobra depois de projetar) é **ortogonal** a (x). Isso é o “ângulo reto” no triângulo da projeção: o segmento que liga (y) ao ponto projetado é perpendicular à reta.

Você pode ver isso diretamente:

$$
\langle x, y-a x\rangle=\langle x,y\rangle-a\langle x,x\rangle=0
\ \Rightarrow\
a=\frac{\langle y,x\rangle}{\langle x,x\rangle}.
$$

**Exemplo** 

Por exemplo, considere o vetor observado: $y=(3,4)$ e o vetor gerador do subespaço (eixo horizontal): $x=(1,0)$. O subespaço gerado por $x$ é o eixo $x$.

Projeção

A projeção de $y$ sobre a reta gerada por $x$ é:

$$
\text{proj}_x(y) = \frac{\langle y,x \rangle}{\|x\|^2} x = (\frac{3.1 + 4.0}{1.1 + 0.0})(1,0) = \frac{3}{1}.(1,0)= (3,0)
$$

**Checagem da ortogonalidade**

* Produto interno: ($\langle y,x\rangle = 3\cdot 1 + 4\cdot 0 = 3$).
* Norma ao quadrado: ($|x|^2 = 1^2+0^2=1$).
* Coeficiente: ($a^*=3/1=3$).
* Projeção: ($\hat y = 3(1,0)=(3,0)$).

E o resíduo é ($r=(3,4)-(3,0)=(0,4)$), que é perpendicular a ($x=(1,0)$) porque ($\langle (1,0),(0,4)\rangle=0$).



Vejamos uma representação gráfica dessa projeção em Python:

```python
import matplotlib.pyplot as plt
import numpy as np

# Define vectors
y = np.array([3, 4])
x = np.array([1, 0])

# Projeção de y em x
proj_y_em_x = (np.dot(y, x) / np.dot(x, x)) * x

# Error (residual) vector
u = y - proj_y_em_x

# Plot
plt.figure()

# Axes: blue and dotted
plt.axhline(0, linestyle=':', color='blue')
plt.axvline(0, linestyle=':', color='blue')

# Plot vectors
plt.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1)

# Projection vector in red
plt.quiver(
    0, 0,
    proj_y_em_x[0], proj_y_em_x[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    linestyle='dashed',
    color='red'
)

# Residual (error) segment: vertical line from (3,0) to (3,4)
plt.plot([y[0], y[0]], [proj_y_em_x[1], y[1]], linestyle='--', color='black')
# Right-angle marker
angle_x = [y[0], y[0] - 0.3, y[0] - 0.3]
angle_y = [proj_y_em_x[1], proj_y_em_x[1], proj_y_em_x[1] + 0.3]
plt.plot(angle_x, angle_y, color='black')

# Labels
plt.text(y[0], y[1], 'y = (3,4)', ha='left', va='bottom')
plt.text(x[0], x[1], 'x = (1,0)', ha='left', va='bottom')
plt.text(proj_y_em_x[0], proj_y_em_x[1], 'Projₓ(y) = (3,0)', ha='left', va='top', color='red')
plt.text(y[0] + 0.05, (y[1] + proj_y_em_x[1]) / 2, 'u', va='center')

# Formatting
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Projeção, erro mínimo e ortogonalidade')

plt.grid(False)
plt.show()
```

* Como ler a figura:
  * A seta que sai da origem até $(3,4)$ é o vetor observado $y$.
  * A reta horizontal é o subespaço gerado por $x=(1,0$).
  * O ponto $(3,0)$ é a projeção ortogonal de $y$ sobre esse subespaço.
  * O segmento vertical ligando $(3,4)$ a $(3,0)$ é o erro mínimo.
  * Esse erro forma um ângulo de $90°$ com o eixo horizontal, logo, $⟨u,x⟩=0$
    * Repare que o $u = y - Projₓ(y)$, ou seja, $(3,4) - (3,0) = (0,4)$. E o produto interno entre $u$ e $Projₓ(y)$ é $0$.

### Momento (*momentum*)

* Em **física**, "momento" (quantidade de movimento) é ($p = m v$).
* Em **mecânica**, “momento” muitas vezes significa **momento de uma força (torque)**: “força × distância ao pivô”.
* Em **probabilidade/estatística/econometria**, “momento” significa **médias de potências** (expectativas) e, mais geral, **condições de momentos** ($E[g(W,\theta)]=0$).



**Momento como “peso × distância” e condição de equilíbrio**

Imagine uma gangorra (alavanca) com um **pivô** no ponto ($c$). Se colocamos massas ($m_i$) em posições ($y_i$) (numa reta), o “efeito de rotação” em torno do pivô cresce quando:

* a massa é maior;
* a distância ao pivô é maior.

Assim, define-se o **momento de primeira ordem em torno de ($c$)** como:

$$
M_1(c)=\sum_{i=1}^n m_i(y_i-c).
$$

* Se ($y_i>c$), o termo é positivo (lado direito).
* Se ($y_i<c$), o termo é negativo (lado esquerdo).

**Condição de equilíbrio** (balanço perfeito em torno do pivô):
$$
M_1(c)=\sum_{i=1}^n m_i(y_i-c)=0.
$$

Resolvendo para ($c$):

$$
\sum_{i=1}^n m_i y_i - c\sum_{i=1}^n m_i = 0
\quad\Rightarrow\quad
\boxed{
c=\frac{\sum_{i=1}^n m_i y_i}{\sum_{i=1}^n m_i}
}
$$

Ou seja: **o pivô que equilibra é a média ponderada das posições** é chamado de **centro de massa**.


**Transformando “massas” em “pesos” (probabilidade)**

Em estatística, é conveniente normalizar as massas para virarem **pesos que somam 1**:

$$
w_i=\frac{m_i}{\sum_{j=1}^n m_j},
\qquad \sum_{i=1}^n w_i=1.
$$

A fórmula do centro de massa vira:

$$
\boxed{
c = \sum_{i=1}^n w_i y_i = \bar y_w
}
$$

E a condição de equilíbrio vira uma identidade simples:

$$
\sum_{i=1}^n w_i(y_i-\bar y_w)=0.
$$

Essa equação já é, na prática, um **exemplo de condição de momento**.


**Centro de massa como esperança de (Y)**

* **Caso discreto (massa de probabilidade)**

Suponha que (Y) seja uma v.a. discreta que pode assumir valores ({y_1,\dots,y_K}), com

$$
p_k = P(Y=y_k), \qquad p_k\ge 0,\qquad \sum_{k=1}^K p_k = 1.
$$

Interprete ($p_k$) como a “massa” colocada no ponto ($y_k$) de uma gangorra (ou uma régua). O **momento de primeira ordem** em torno de ($c$) é
$$
M_1(c)=\sum_{k=1}^K p_k (y_k-c).
$$

A **condição de equilíbrio** é ($M_1(c)=0$), isto é,
$$
\sum_{k=1}^K p_k(y_k-c)=0
\Rightarrow
\sum_{k=1}^K p_k y_k - c\sum_{k=1}^K p_k=0
\Rightarrow
\boxed{
c=\sum_{k=1}^K p_k y_k = E[Y].
}
$$

Ou seja: **o centro de massa da distribuição discreta é a esperança ($E[Y]$)**.

* ***Análogo amostral (distribuição empírica)***

Se você tem uma amostra ($y_1,\dots,y_n$), a distribuição empírica atribui massa ($1/n$) a cada observação. Então
$$
\boxed{
\bar y = \frac{1}{n}\sum_{i=1}^n y_i
}
$$

é o “centro de massa” da amostra, pois

$$
\sum_{i=1}^n \frac{1}{n}(y_i-\bar y)=0
$$

* **Caso contínuo (densidade como “massa por unidade de comprimento”)**

Se ($Y$) é contínua com densidade ($f_Y(y)$) (com ($\int f_Y(y)dy=1)$), o momento em torno de ($c$) é

$$
M_1(c)=\int (y-c) f_Y(y)dy.
$$

Equilíbrio ($M_1(c)=0$) implica

$$
\boxed{
c=\int y f_Y(y)dy = E[Y].
}
$$



***Momentos estatísticos: “média de potências”***

Agora vem a generalização natural: podemos medir elevada a potências.

* **Momento de ordem ($k$) (em torno da origem)**

Para uma variável aleatória ($Y$), o **momento bruto (raw moment)** de ordem ($k$) é:

$$
\mu_k' = E[Y^k].
$$

A versão amostral (análogo amostral/empírico) é:
$$
m_k'=\frac{1}{n}\sum_{i=1}^n y_i^k.
$$

Se houver pesos ($w_i$):
$$
m_{k,w}'=\sum_{i=1}^n w_i y_i^k.
$$


* **Momento central de ordem ($k$) (em torno da média)**

O **momento central** de ordem ($k$) é:

$$
\mu_k = E\big[(Y-E[Y])^k\big].
$$

A versão amostral (empírica) é:

$$
m_k=\frac{1}{n}\sum_{i=1}^n (y_i-\bar y)^k
\quad
(\text{ou ponderada: } \sum w_i (y_i-\bar y_w)^k).
$$

Casos fundamentais:

* ($k=1$): ($\mu_1=0$) (equilíbrio em torno da média).
* ($k=2$): ($\mu_2=\mathrm{Var}(Y)$) (dispersão).
* ($k=3$): ($\mu_3$) (assimetria - skewness).
* ($k=4$): ($\mu_4$) (curtose - kurtosis).



**Condição de Momento**

Em econometria, as **condições de momento** são equações que relacionam as variáveis observáveis e os parâmetros desconhecidos do modelo. Elas são usadas para estimar os parâmetros do modelo a partir dos dados. São uma generalização das condições de equilíbrio, onde a função $g(W,\theta)$ pode ser mais complexa do que apenas $y-c$.

Uma **condição de momento** é uma equação do tipo:

$$E[g(W,\theta)]=0$$

* $W$ é um vetor de variáveis aleatórias observáveis, 
* $\theta$ é um vetor de parâmetros desconhecidos, e 
* $g$ é uma função que relaciona as variáveis observáveis e os parâmetros.


Vamos passo a passo.

**Modelo econométrico estrutural e a Condição de Momento**

Na econometria padrão, partimos de um modelo estrutural para a variável de interesse:

$$
Y = f(X) + \varepsilon,
$$

onde:

* ($Y$) é a variável resposta,
* ($X$) é o vetor de covariáveis,
* ($\varepsilon$) é o termo de erro (fatores não observados).

Sob a hipótese de linearidade:

$$
Y = X'\beta + \varepsilon.
$$

Certo, essa é a forma usual do OLS. Vamos 
Definimos o vetor de variáveis observadas como:

$$
W = (Y, D, X),
$$

onde, no contexto causal:

* ($Y$): outcome,
* ($D$): tratamento,
* ($X$): covariáveis.

Uma observação individual é:

$$
W_i = (Y_i, D_i, X_i),
$$

isto é, um ponto no espaço amostral.

Assumimos que existe uma ***distribuição conjunta verdadeira***:

$$
P = \mathcal{L}(Y, D, X),
$$

isto é, a lei de probabilidade de ($W$). Logo, escrevemos:

$$
W \sim P
$$

significando que ($W$) é gerado segundo a distribuição ($P$).


temos uma **amostra aleatória** quando observamos uma amostra i.i.d.:

$$
W_1, W_2, \ldots, W_n \sim P,
$$

onde cada observação é um sorteio independente da **mesma** distribuição populacional.


* **Parâmetro como funcional de (P)**

Em econometria, o objeto de interesse quase sempre pode ser escrito como um **funcional** da distribuição:
$$
\theta_0=\Theta(P).
$$

Exemplos:

* média: ($\Theta(P)=E_P[Y]$);
* variância: ($\Theta(P)=E_P[(Y-E_P[Y])^2]$);
* regressão linear (sob exogeneidade): ($\Theta(P)$) resolve ($E_P[X(Y-X'\beta)]=0$).


**Condições de momento (forma mais geral)**

Muitos parâmetros são definidos como solução de uma **equação de momento populacional**:
$$
E_P[\psi(W,\theta_0)]=0.
$$

Exemplos:

* **média**: $(\psi(W,\theta)=Y-\theta) (\Rightarrow) (E_P[Y-\theta_0]=0)$.
* **OLS**: $(\psi(W,\beta)=X(Y-X'\beta)) (\Rightarrow) (E_P[X(Y-X'\beta_0)]=0)$.

**como (P) é desconhecida, não conseguimos computar ($E_P[\cdot]$)**.



A **distribuição empírica** é o objeto que torna a média amostral uma “integral”, espelhando (P):
$$
\hat P_n ;=; \frac{1}{n}\sum_{i=1}^n \delta_{W_i}.
$$

Intuição (sem Dirac): (\hat P_n) **coloca massa (1/n) em cada ponto observado** (W_i).

A propriedade-chave é:
$$
\int g(w), d\hat P_n(w)
;=;
\frac{1}{n}\sum_{i=1}^n g(W_i).
$$

> Ou seja: a distribuição empírica não é “uma ideia nova”; ela é só uma forma elegante de escrever “média amostral”.



### 5. Distribuição empírica

A distribuição verdadeira ($P$) é desconhecida. Nós a aproximamos pela **distribuição empírica**:

$$
\hat P_n = \frac{1}{n} \sum_{i=1}^n \delta_{W_i},
$$

onde:

* ($\delta_{W_i}$) é a **medida de Dirac** no ponto ($W_i$),
* Ela atribui massa 1 ao ponto ($W_i$) e 0 aos demais.

Formalmente, para qualquer conjunto mensurável (A):

$$
\delta_{W_i}(A) =
\begin{cases}
1, & \text{se } W_i \in A, \
0, & \text{se } W_i \notin A.
\end{cases}
$$

Logo, ($\hat P_n(A)$) é simplesmente a **proporção da amostra que cai em (A)**.

### 6. Interpretação fundamental

A lógica central da econometria (e da inferência semiparamétrica) é:

$$
P ;; \text{(desconhecida)}
\quad \longrightarrow \quad
\hat P_n ;; \text{(aproximação empírica)}
$$

Todo parâmetro que estimamos — média, regressão, ATE, ATT, CATE — será escrito como:

$$
\theta_0 = \Theta(P),
\qquad
\hat\theta_n = \Theta(\hat P_n).
$$

Ou seja:

> Primeiro vem a distribuição.
> Depois vem o funcional.
> Depois vem o estimador plug-in.

Esse é o alicerce sobre o qual depois entram:

* identificação,
* função de influência,
* eficiência,
* DML,
* DAGs e causalidade.

Concordo: do jeito que está, a “distribuição empírica” entra **como um objeto novo (medida de Dirac)** sem você ter dito antes **por que** precisamos dela. O conserto é inserir **duas pontes** antes do item (5):

1. **Expectativa = integral sob (P)** (para qualquer função (g(W)));
2. **Princípio do análogo amostral**: trocar (E_P[\cdot]) por média amostral (\frac1n\sum).

A distribuição empírica aparece então só como uma **forma compacta** de escrever essa troca.

## Do modelo estrutural à condição de momento (sem salto)

### 1) DGP, dados observados e distribuição

Assumimos que existe uma **lei populacional** (desconhecida) que gera os dados:
$$
W=(Y,D,X)\sim P.
$$

Uma amostra i.i.d. significa:
$$
W_1,\dots,W_n \ \text{i.i.d.}\ \sim P.
$$

> Interpretação: existe uma “urna” populacional (P). A amostra são (n) sorteios independentes dessa mesma urna.


### 2) Parâmetro como funcional de (P)

Em econometria, o objeto de interesse quase sempre pode ser escrito como um **funcional** da distribuição:
$$
\theta_0=\Theta(P).
$$

Exemplos:

* média: (\Theta(P)=E_P[Y]);
* variância: (\Theta(P)=E_P[(Y-E_P[Y])^2]);
* regressão linear (sob exogeneidade): (\Theta(P)) resolve (E_P[X(Y-X'\beta)]=0).

### 3) Condições de momento (forma mais geral)

Muitos parâmetros são definidos como solução de uma **equação de momento populacional**:
$$
E_P[\psi(W,\theta_0)]=0.
$$

Exemplos:

* **média**: (\psi(W,\theta)=Y-\theta) (\Rightarrow) (E_P[Y-\theta_0]=0).
* **OLS**: (\psi(W,\beta)=X(Y-X'\beta)) (\Rightarrow) (E_P[X(Y-X'\beta_0)]=0).

Isso já cria o “gancho” natural para o estimador: **como (P) é desconhecida, não conseguimos computar (E_P[\cdot])**.

### 4) Ponte essencial: expectativa como integral

Para qualquer função mensurável (g(W)), a expectativa sob (P) pode ser escrita como uma integral:
$$
E_P[g(W)] ;=; \int g(w), dP(w).
$$

Essa frase é o ponto onde “probabilidade” vira “objeto matemático manipulável”.

---

### 5) Princípio do análogo amostral (a regra de estimação)

Como (P) é desconhecida, substituímos a expectativa populacional por sua contraparte amostral:
$$
E_P[g(W)] \approx \frac{1}{n}\sum_{i=1}^n g(W_i).
$$

Aplicando isso na condição de momento:
$$
E_P[\psi(W,\theta_0)]=0
\quad\leadsto\quad
\frac{1}{n}\sum_{i=1}^n \psi(W_i,\hat\theta)=0.
$$

Esse é o “motor” por trás de:

* método dos momentos,
* OLS (normal equations),
* GMM,
* M-estimação, etc.

### 6) Agora sim: distribuição empírica (só para compactar a notação)

A **distribuição empírica** é o objeto que torna a média amostral uma “integral”, espelhando (P):
$$
\hat P_n ;=; \frac{1}{n}\sum_{i=1}^n \delta_{W_i}.
$$

Intuição (sem Dirac): (\hat P_n) **coloca massa (1/n) em cada ponto observado** (W_i).

A propriedade-chave é:
$$
\int g(w), d\hat P_n(w)
;=;
\frac{1}{n}\sum_{i=1}^n g(W_i).
$$

> Ou seja: a distribuição empírica não é “uma ideia nova”; ela é só uma forma elegante de escrever “média amostral”.

Com isso, sua frase central fica natural (sem salto):

\qquad
\hat\theta_n=\Theta(\hat P_n).
$$

E, no caso de momentos:
$$
E_P[\psi(W,\theta_0)]=0
\quad\Longrightarrow\quad
E_{\hat P_n}[\psi(W,\hat\theta)]=0.
$$

## Centro de massa como esperança de (Y)

### Caso discreto (massa de probabilidade)

Suponha que (Y) seja uma v.a. discreta que pode assumir valores ({y_1,\dots,y_K}), com
[
p_k = P(Y=y_k), \qquad p_k\ge 0,\qquad \sum_{k=1}^K p_k = 1.
]

Interprete (p_k) como a “massa” colocada no ponto (y_k) de uma régua. O **momento de primeira ordem** em torno de (c) é
[
M_1(c)=\sum_{k=1}^K p_k (y_k-c).
]

A **condição de equilíbrio** é (M_1(c)=0), isto é,
[
\sum_{k=1}^K p_k(y_k-c)=0
;\Rightarrow;
\sum_{k=1}^K p_k y_k - c\sum_{k=1}^K p_k=0
;\Rightarrow;
\boxed{
c=\sum_{k=1}^K p_k y_k = E[Y].
}
]

Ou seja: **o centro de massa da distribuição discreta é a esperança (E[Y])**.

### Análogo amostral (distribuição empírica)

Se você tem uma amostra (y_1,\dots,y_n), a distribuição empírica atribui massa (1/n) a cada observação. Então
[
\boxed{
\bar y = \frac{1}{n}\sum_{i=1}^n y_i
}
]
é o “centro de massa” da amostra, pois
[
\sum_{i=1}^n \frac{1}{n}(y_i-\bar y)=0.
]

### Caso contínuo (densidade como “massa por unidade de comprimento”)

Se (Y) é contínua com densidade (f_Y(y)) (com (\int f_Y(y),dy=1)), o momento em torno de (c) é
[
M_1(c)=\int (y-c), f_Y(y),dy.
]
Equilíbrio (M_1(c)=0) implica
[
\boxed{
c=\int y, f_Y(y),dy = E[Y].
}
]

### Modelo econométrico estrutural

Na econometria padrão, partimos de um modelo estrutural para a variável de interesse:

$$
Y = f(X) + \varepsilon,
$$

onde:

* ($Y$) é a variável resposta,
* ($X$) é o vetor de covariáveis,
* ($\varepsilon$) é o termo de erro (fatores não observados).

Sob a hipótese de linearidade:

$$
Y = X'\beta + \varepsilon.
$$

### 2. Vetor de observáveis

Definimos o vetor de variáveis observadas como:

$$
W = (Y, D, X),
$$

onde, no contexto causal:

* ($Y$): outcome,
* ($D$): tratamento,
* ($X$): covariáveis.

Uma observação individual é:

$$
W_i = (Y_i, D_i, X_i),
$$

isto é, um ponto no espaço amostral.

### 3. Distribuição populacional

Assumimos que existe uma distribuição conjunta verdadeira:

$$
P = \mathcal{L}(Y, D, X),
$$

isto é, a lei de probabilidade de ($W$).

Escrevemos:

$$
W \sim P,
$$

significando que ($W$) é gerado segundo a distribuição ($P$).


### 4. Amostra aleatória

Observamos uma amostra i.i.d.:

$$
W_1, W_2, \ldots, W_n \sim P,
$$

onde cada observação é um sorteio independente da **mesma** distribuição populacional.


### 5. Distribuição empírica

A distribuição verdadeira ($P$) é desconhecida. Nós a aproximamos pela **distribuição empírica**:

$$
\hat P_n = \frac{1}{n} \sum_{i=1}^n \delta_{W_i},
$$

onde:

* ($\delta_{W_i}$) é a **medida de Dirac** no ponto ($W_i$),
* Ela atribui massa 1 ao ponto ($W_i$) e 0 aos demais.

Formalmente, para qualquer conjunto mensurável (A):

$$
\delta_{W_i}(A) =
\begin{cases}
1, & \text{se } W_i \in A, \
0, & \text{se } W_i \notin A.
\end{cases}
$$

Logo, ($\hat P_n(A)$) é simplesmente a **proporção da amostra que cai em (A)**.

### 6. Interpretação fundamental

A lógica central da econometria (e da inferência semiparamétrica) é:

$$
P ;; \text{(desconhecida)}
\quad \longrightarrow \quad
\hat P_n ;; \text{(aproximação empírica)}
$$

Todo parâmetro que estimamos — média, regressão, ATE, ATT, CATE — será escrito como:

$$
\theta_0 = \Theta(P),
\qquad
\hat\theta_n = \Theta(\hat P_n).
$$

Ou seja:

> Primeiro vem a distribuição.
> Depois vem o funcional.
> Depois vem o estimador plug-in.

Esse é o alicerce sobre o qual depois entram:

* identificação,
* função de influência,
* eficiência,
* DML,
* DAGs e causalidade.

**Estimação por condições de momento**

A econometria quase sempre estima parâmetros impondo que certos ***momentos populacionais sejam zero***.

* **Exemplo 1** — A média como solução de um momento

Queremos ($\theta=E[Y]$). Isso equivale a:

$$
E[Y-\theta]=0.
$$

O análogo amostral resolve:

$$
\frac{1}{n}\sum_{i=1}^n (y_i-\theta)=0
\quad\Rightarrow\quad
\boxed{\hat\theta=\bar y}
$$

Aqui, a “condição de equilíbrio” ($\sum (y_i-\bar y)=0$) aparece exatamente como **momento amostral igual a zero**.

* **Exemplo 2** — OLS como projeção e condição de momento

No modelo linear ($Y=X\beta+u$), a condição populacional chave é:
$$
E[X'u]=0
\quad\text{(em componentes: } E[X_j u]=0 \text{ para cada } j).
$$

O análogo amostral é:
$$
X'(y-X\beta)=0
\quad\Rightarrow\quad
\hat\beta=(X'X)^{-1}X'y.
$$

Isso é exatamente a mesma ideia geométrica que você já explicou:

* a projeção minimiza distância;
* o resíduo fica ortogonal ao espaço gerado por (X);
* “ortogonalidade” vira “momento igual a zero”.

## Introdução à Análise Causal: Identificação, Estimação e Inferência

### Identificação Causal

Identificação causal é a ponte entre o modelo teórico de causalidade e a distribuição dos dados observados. Trata-se do estudo das condições sob as quais um parâmetro causal — definido em termos de resultados potenciais — pode ser escrito como um **funcional único da distribuição observável**.

Formalmente, um parâmetro causal ($\theta_0$) é identificado se existe um **funcional** ($\Theta : \mathcal P \to \mathbb R$) tal que

$$
\theta_0 = \Theta(P_W),
$$


onde ($P_W$) é a distribuição conjunta das variáveis observadas. Assim, identificação não é uma propriedade do estimador nem da amostra, mas do **modelo populacional sob hipóteses estruturais**.

Em termos intuitivos, identificação responde à pergunta:

> Sob quais hipóteses teóricas o contrafactual não observado pode ser recuperado a partir dos dados observáveis?

Aqui entram:

* **Representações causais**:

  * Resultados potenciais, DAGs, critérios de back-door e front-door.
* **Hipóteses estruturais**:

  * Exogeneidade, ignorabilidade condicional, overlap, paralelismo, continuidade, exclusão, monotonicidade.
* **Parâmetros causais**:

  * ATE, ATT, ATU, CATE, LATE, efeitos dinâmicos, efeitos de coorte.

#### Estratégias (Designs) de Identificação

Os chamados *designs* ou *quase-experimentos* não são estimadores, mas **conjuntos de hipóteses que tornam o parâmetro causal identificável**. Eles especificam como o contrafactual é recuperado:

* **Diferença-em-Diferenças (DiD):** identificação via tendências paralelas.
* **Regression Discontinuity (RD):** identificação via continuidade no ponto de corte.
* **Variáveis Instrumentais (IV):** identificação via exogeneidade, relevância e exclusão.
* **Synthetic Control (SC):** identificação via combinação convexa que reproduz a trajetória contrafactual.

Perguntas típicas no nível da identificação:

* O efeito causal é, em princípio, recuperável a partir da distribuição observada?
* Quais hipóteses tornam o contrafactual bem definido?
* Minha estratégia gera um contrafactual teoricamente crível?

### Estimação Causal

Dada a identificação, a estimação causal trata de **como aproximar, em amostra finita, o funcional identificador ($\Theta(P)$)**.

Ou seja, se

$$
\theta_0 = \Theta(P),
$$

buscamos um estimador ($\hat\theta_n = \Theta(\hat P_n)$) que seja consistente e, idealmente, eficiente.

Aqui entram as **técnicas econométricas**, que podem ser organizadas segundo a forma como operacionalizam o funcional:

1. **Imputação (Outcome Regression):**

   * Regressões condicionais, Matching, PSM, TWFE, modelos estruturais.
2. **Reponderação (Weighting):**

   * IPW, entropy balancing, Synthetic Control, balancing weights.
3. **Equações de Momento / Ortogonalização:**

   * IV, GMM, DID2S, DR, DML, TMLE.

Um mesmo design de identificação pode ser implementado por estimadores distintos:

* **DiD:** TWFE (imputação), Synthetic DiD (ponderação), DID2S / DR-DiD (momentos).
* **RDD:** regressão local (imputação), kernel weighting (ponderação).
* **IV:** 2SLS (momentos), control function (imputação), LIML.

Perguntas no nível da estimação:

* Qual estimador computa corretamente o funcional identificado?
* O estimador é consistente e eficiente?
* Como lidar com viés de pequena amostra, regularização e aprendizado de máquina?

### Inferência Causal

A inferência causal trata da **distribuição amostral do estimador** e da quantificação da incerteza sobre o parâmetro causal.

Formalmente, estuda a lei assintótica (ou finita) de:

$$
\sqrt{n}(\hat\theta_n - \theta_0),
$$

permitindo construir erros-padrão, intervalos de confiança e testes de hipóteses.

Inclui:

* Matrizes de variância robustas, clusterizadas e HAC.
* Bootstrap (clássico, wild cluster, multiplier).
* Inferência baseada em função de influência e eficiência semiparamétrica.
* Bandas uniformes para efeitos dinâmicos (event studies).

Perguntas típicas:

* Qual é a variância assintótica do estimador?
* Posso rejeitar a hipótese de efeito nulo?
* Quão precisa é a estimativa do efeito causal?

Em suma:

> Identificação diz **o que** pode ser aprendido dos dados;
> Estimação diz **como** calcular esse objeto;
> Inferência diz **com que precisão** ele é conhecido.
