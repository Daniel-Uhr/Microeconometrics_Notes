## Fundamentos Matemáticos

### Definições

* **Geometria**: estudo de formas, tamanhos, posições relativas de figuras e das propriedades do espaço.

### Conjuntos Numéricos Importantes

* **$\mathbb{N}$**: conjunto dos números naturais: $\{1, 2, 3, \ldots\}$.
  * Exemplo: contagem de objetos — 1 maçã, 2 laranjas, 3 bananas.

* **$\mathbb{Z}$**: conjunto dos números inteiros: $\{\ldots, -3, -2, -1, 0, 1, 2, 3, \ldots\}$.
  * Exemplo: temperaturas abaixo e acima de zero — -5°C, 0°C, 10°C.

* **$\mathbb{Q}$**: conjunto dos números racionais, que podem ser escritos como fração $\frac{a}{b}$, com $a, b \in \mathbb{Z}$ e $b \neq 0$.
  * Exemplo: $\frac{1}{2}$, $\frac{3}{4}$, $\frac{-5}{2}$.

* **$\mathbb{I}$**: conjunto dos números irracionais, que não podem ser escritos como fração de inteiros.
  * Exemplo: $\sqrt{2}$, $\pi$, $e$.

* **$\mathbb{R}$**: conjunto dos números reais, que inclui os racionais e irracionais.
  * Exemplo: medidas contínuas — altura (1.75 m), peso (70.5 kg).

* **$\mathbb{C}$**: conjunto dos números complexos, da forma $a + bi$, com $a,b \in \mathbb{R}$ e $i^2=-1$.
  * Exemplo: $3+4i$, $-2-5i$.
  * Aplicações em engenharia elétrica e física.

Existem muitos outros conjuntos numéricos, mas **para a econometria**, o conjunto fundamental é o dos **números reais** e os **espaços vetoriais construídos a partir deles**.


## Espaços Vetoriais e Geometria dos Dados

### Espaço Vetorial

Um **espaço vetorial** é um conjunto de objetos, chamados vetores, que podem ser:
- somados entre si;
- multiplicados por escalares;

obedecendo a propriedades como associatividade, comutatividade, existência de elemento neutro e de inverso aditivo.


### O Espaço $\mathbb{R}^k$

O conjunto $\mathbb{R}^k$ não é apenas um conjunto: ele é um **espaço vetorial real de dimensão $k$**.

Cada elemento de $\mathbb{R}^k$ é um vetor:
$$
x = (x_1, x_2, \ldots, x_k)
$$

O espaço admite:
- soma de vetores;
- multiplicação por escalares reais.

Casos particulares:
- $\mathbb{R}^1$: a reta real;
- $\mathbb{R}^2$: o plano cartesiano;
- $\mathbb{R}^3$: o espaço tridimensional;
- $\mathbb{R}^k$, com $k>3$: espaço de dimensão maior, não visualizável, mas perfeitamente bem definido matematicamente.



### Produto Interno

Em $\mathbb{R}^k$, define-se o **produto interno euclidiano** entre dois vetores $x,y \in \mathbb{R}^k$ por:
$$
\langle x,y\rangle = \sum_{j=1}^k x_j y_j
$$

Interpretação geométrica (soma de produtos coordenada a coordenada):
- o produto interno mede o **grau de alinhamento** entre dois vetores;
- é máximo quando os vetores são paralelos;
- é zero quando os vetores são ortogonais (ângulo de 90°). 
- Representa independência direcional.
  - obs: a covariância é um produto interno aplicado aos vetores centrados na média, dividido por *n* para ter interpretação probabilística 
    - $\operatorname{cov}(X,Y) = \frac{1}{n} \sum_{i=1}^n (X_i - \bar{X})(Y_i - \bar{Y})$ = $\frac{1}{n} \langle X - \bar{X}, Y - \bar{Y} \rangle$.

### Norma (Comprimento do Vetor)

A **norma euclidiana** de um vetor $x \in \mathbb{R}^k$ é definida por:
$$
\|x\| = \sqrt{\sum_{j=1}^k x_j^2}
$$

- a norma transforma um vetor em um número real não negativo;
- geometricamente, ela representa o **comprimento da seta que liga a origem ao ponto $x$**.
- OBS: a norma tem relação direta com o desvio padrão de um vetor centrado (e com a variância, que é o quadrado da norma de um vetor centrado dividido por *n*). 

***Exemplo aplicado:***
$$
x=(3,4)\in\mathbb{R}^2 \quad \Rightarrow \quad \|x\|=\sqrt{3^2+4^2}=5
$$

### Métrica (Distância entre Pontos)

A norma induz uma **métrica**, isto é, uma noção de distância entre dois pontos $x,y \in \mathbb{R}^k$:
$$
d(x,y)=\|x-y\|
$$

Exemplo simples:
- no espaço $\mathbb{R}$, a métrica usual é:
$$
d(x,y)=|x-y|
$$


### Projeção de Vetores

Vamos entender a projeção de vetores porque ela é fundamental para a compreensão de conceitos como **regressão linear** e **ajuste de modelos**.

O subespaço gerado por ($x\neq 0$) é a reta

$$
\mathcal{L}={a x: a\in\mathbb{R}}.
$$

Projetar ($y$) nessa reta significa achar o ponto da reta ($ax$) **mais próximo** de ($y$) (em norma Euclidiana). Ou seja, resolver:

$$
\min_{a\in\mathbb{R}}\ |y-ax|^2.
$$

Obs: o quadrado não muda o minimizador, só muda a escala da função objetivo, e é mais fácil de derivar.

Chame o minimizador de ($a^{*}$). Então a projeção é
$$
\operatorname{proj}_x(y)=a^*x.
$$

***Derivação por minimização***

Expanda a função objetivo:

$$
|y-ax|^2=\langle y-ax,\ y-ax\rangle
= \langle y,y\rangle -2a\langle y,x\rangle + a^2\langle x,x\rangle.
$$

Isso é uma parábola em ($a$). Derive e iguale a zero:

$$
\frac{d}{da}|y-ax|^2 = -2\langle y,x\rangle +2a\langle x,x\rangle=0.
$$

Logo,

$$
a^*=\frac{\langle y,x\rangle}{\langle x,x\rangle}
=\frac{\langle y,x\rangle}{|x|^2}.
$$

Portanto,

$$
\boxed{\operatorname{proj}_x(y)=\frac{\langle y,x\rangle}{|x|^2} x}
$$

Essa é exatamente a fórmula da projeção de um vetor sobre outro.


***Interpretação geométrica: o “erro” é perpendicular***

Defina

$$
\hat y=\operatorname{proj}_x(y)=a^*x,\qquad r=y-\hat y.
$$

A condição de otimalidade acima é equivalente a:

$$
\langle x,\ r\rangle =0.
$$

Ou seja, o **resíduo** ($r$) (a parte de ($y$) que sobra depois de projetar) é **ortogonal** a ($x$). Isso é o “ângulo reto” no triângulo da projeção: o segmento que liga ($y$) ao ponto projetado é perpendicular à reta.

Você pode ver isso diretamente:

$$
\langle x, y-a x\rangle=\langle x,y\rangle-a\langle x,x\rangle=0
\ \Rightarrow\
a=\frac{\langle y,x\rangle}{\langle x,x\rangle}.
$$

**Exemplo** 

Considere o vetor observado: $y=(3,4)$ e o vetor gerador do subespaço (eixo horizontal): $x=(1,0)$. O subespaço gerado por $x$ é o eixo $x$.

A projeção de $y$ sobre a reta gerada por $x$ é:

$$
\text{proj}_x(y) = \frac{\langle y,x \rangle}{\|x\|^2} x = (\frac{3.1 + 4.0}{1.1 + 0.0})(1,0) = \frac{3}{1}.(1,0)= (3,0)
$$

***Checagem da ortogonalidade***

* Produto interno: ($\langle y,x\rangle = 3\cdot 1 + 4\cdot 0 = 3$).
* Norma ao quadrado: ($|x|^2 = 1^2+0^2=1$).
* Coeficiente: ($a^*=3/1=3$).
* Projeção: ($\hat y = 3(1,0)=(3,0)$).

E o resíduo é ($r=(3,4)-(3,0)=(0,4)$), que é perpendicular a ($x=(1,0)$) porque ($\langle (1,0),(0,4)\rangle=0$).



Vejamos uma representação gráfica dessa projeção em Python:

```python
import matplotlib.pyplot as plt
import numpy as np

# Define vectors
y = np.array([3, 4])
x = np.array([1, 0])

# Projeção de y em x
proj_y_em_x = (np.dot(y, x) / np.dot(x, x)) * x

# Error (residual) vector
u = y - proj_y_em_x

# Plot
plt.figure()

# Axes: blue and dotted
plt.axhline(0, linestyle=':', color='blue')
plt.axvline(0, linestyle=':', color='blue')

# Plot vectors
plt.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1)

# Projection vector in red
plt.quiver(
    0, 0,
    proj_y_em_x[0], proj_y_em_x[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    linestyle='dashed',
    color='red'
)

# Residual (error) segment: vertical line from (3,0) to (3,4)
plt.plot([y[0], y[0]], [proj_y_em_x[1], y[1]], linestyle='--', color='black')
# Right-angle marker
angle_x = [y[0], y[0] - 0.3, y[0] - 0.3]
angle_y = [proj_y_em_x[1], proj_y_em_x[1], proj_y_em_x[1] + 0.3]
plt.plot(angle_x, angle_y, color='black')

# Labels
plt.text(y[0], y[1], 'y = (3,4)', ha='left', va='bottom')
plt.text(x[0], x[1], 'x = (1,0)', ha='left', va='bottom')
plt.text(proj_y_em_x[0], proj_y_em_x[1], 'Projₓ(y) = (3,0)', ha='left', va='top', color='red')
plt.text(y[0] + 0.05, (y[1] + proj_y_em_x[1]) / 2, 'u', va='center')

# Formatting
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Projeção, erro mínimo e ortogonalidade')

plt.grid(False)
plt.show()
```

* Como ler a figura:
  * A seta que sai da origem até $(3,4)$ é o vetor observado $y$.
  * A reta horizontal é o subespaço gerado por $x=(1,0$).
  * O ponto $(3,0)$ é a projeção ortogonal de $y$ sobre esse subespaço.
  * O segmento vertical ligando $(3,4)$ a $(3,0)$ é o erro mínimo.
  * Esse erro forma um ângulo de $90°$ com o eixo horizontal, logo, $⟨u,x⟩=0$
    * Repare que o $u = y - Projₓ(y)$, ou seja, $(3,4) - (3,0) = (0,4)$. E o produto interno entre $u$ e $Projₓ(y)$ é $0$.

### Momento (*momentum*)

* Em **física**, "momento" (quantidade de movimento) é ($p = m v$).
* Em **mecânica**, “momento” muitas vezes significa **momento de uma força (torque)**: “força × distância ao pivô”.
* Em **probabilidade/estatística/econometria**, “momento” significa **médias de potências** (expectativas) e, mais geral, **condições de momentos** ($E[g(W,\theta)]=0$).



**Momento como “peso × distância” e condição de equilíbrio**

Imagine uma gangorra (alavanca) com um **pivô** no ponto ($c$). Se colocamos massas ($m_i$) em posições ($y_i$) (numa reta), o “efeito de rotação” em torno do pivô cresce quando:

* a massa é maior;
* a distância ao pivô é maior.

Assim, define-se o **momento de primeira ordem em torno de ($c$)** como:

$$
M_1(c)=\sum_{i=1}^n m_i(y_i-c).
$$

* Se ($y_i>c$), o termo é positivo (lado direito).
* Se ($y_i<c$), o termo é negativo (lado esquerdo).

**Condição de equilíbrio** (balanço perfeito em torno do pivô):
$$
M_1(c)=\sum_{i=1}^n m_i(y_i-c)=0.
$$

Resolvendo para ($c$):

$$
\sum_{i=1}^n m_i y_i - c\sum_{i=1}^n m_i = 0
\quad\Rightarrow\quad
\boxed{
c=\frac{\sum_{i=1}^n m_i y_i}{\sum_{i=1}^n m_i}
}
$$

Ou seja: **o pivô que equilibra é a média ponderada das posições** é chamado de **centro de massa**.


**Transformando “massas” em “pesos” (probabilidade)**

Em estatística, é conveniente normalizar as massas para virarem **pesos que somam 1**:

$$
w_i=\frac{m_i}{\sum_{j=1}^n m_j},
\qquad \sum_{i=1}^n w_i=1.
$$

A fórmula do centro de massa vira:

$$
\boxed{
c = \sum_{i=1}^n w_i y_i = \bar y_w
}
$$

E a **condição de equilíbrio** vira uma identidade simples:

$$
\sum_{i=1}^n w_i(y_i-\bar y_w)=0.
$$

Essa equação já é, na prática, um **exemplo de condição de momento**.


**Centro de massa como esperança de ($Y$)**

* **Caso discreto (massa de probabilidade)**

Suponha que ($Y$) seja uma v.a. discreta que pode assumir valores (${y_1,\dots,y_K}$), com

$$
p_k = P(Y=y_k), \qquad p_k\ge 0,\qquad \sum_{k=1}^K p_k = 1.
$$

Interprete ($p_k$) como a “massa” colocada no ponto ($y_k$) de uma gangorra (ou uma régua). O **momento de primeira ordem** em torno de ($c$) é
$$
M_1(c)=\sum_{k=1}^K p_k (y_k-c).
$$

A **condição de equilíbrio** é ($M_1(c)=0$), isto é,
$$
\sum_{k=1}^K p_k(y_k-c)=0
\Rightarrow
\sum_{k=1}^K p_k y_k - c\sum_{k=1}^K p_k=0
\Rightarrow
\boxed{
c=\sum_{k=1}^K p_k y_k = E[Y].
}
$$

Ou seja: **o centro de massa da distribuição discreta é a esperança ($E[Y]$)**.




* ***Análogo amostral (distribuição empírica)***

Se você tem uma amostra ($y_1,\dots,y_n$), a distribuição empírica atribui massa de probabilidade ($1/n$) a cada observação. Então
$$
\boxed{
\bar y = \frac{1}{n}\sum_{i=1}^n y_i
}
$$

é o “centro de massa” da amostra, pois

$$
\sum_{i=1}^n \frac{1}{n}(y_i-\bar y)=0
$$

* **Caso contínuo (densidade como “massa por unidade de comprimento”)**

Se ($Y$) é contínua com densidade ($f_Y(y)$) (com ($\int f_Y(y)dy=1)$), o momento em torno de ($c$) é

$$
M_1(c)=\int (y-c) f_Y(y)dy.
$$

Equilíbrio ($M_1(c)=0$) implica

$$
\boxed{
c=\int y f_Y(y)dy = E[Y].
}
$$



***Momentos estatísticos: “média de potências”***

Agora vem a generalização natural: podemos medir elevada a potências.

* **Momento de ordem ($k$) (em torno da origem)**

Para uma variável aleatória ($Y$), o **momento bruto (raw moment)** de ordem ($k$) é:

$$
\mu_k' = E[Y^k].
$$

A versão amostral (análogo amostral/empírico) é:
$$
m_k'=\frac{1}{n}\sum_{i=1}^n y_i^k.
$$

Se houver pesos ($w_i$):
$$
m_{k,w}'=\sum_{i=1}^n w_i y_i^k.
$$


* **Momento central de ordem ($k$) (em torno da média)**

O **momento central** de ordem ($k$) é:

$$
\mu_k = E\big[(Y-E[Y])^k\big].
$$

A versão amostral (empírica) é:

$$
m_k=\frac{1}{n}\sum_{i=1}^n (y_i-\bar y)^k
\quad
(\text{ou ponderada: } \sum w_i (y_i-\bar y_w)^k).
$$

Casos fundamentais:

* ($k=1$): ($\mu_1=0$) (equilíbrio em torno da média).
* ($k=2$): ($\mu_2=\mathrm{Var}(Y)$) (dispersão).
* ($k=3$): ($\mu_3$) (assimetria - skewness).
* ($k=4$): ($\mu_4$) (curtose - kurtosis).



### Condição de Momento

Em econometria, **condições de momento** são equações que relacionam variáveis observáveis e parâmetros desconhecidos do modelo. Elas generalizam as “condições de equilíbrio” (como no centro de massa), permitindo que a função que entra no momento seja bem mais complexa do que apenas $(y-c)$.

Uma **condição de momento populacional** tem a forma:

$$
E_{P}[\psi(W,\theta_0)]=0,
$$

onde:

- $W$ é o vetor de variáveis aleatórias observáveis (por exemplo, $(Y,D,X)$, ou apenas $(Y,X)$),
  - $W_i=(Y_i,D_i,X_i)$ é a observação $i$ de $W$;
- $\theta_0$ é o parâmetro verdadeiro (fixo na população, mas desconhecido);
- $\psi(\cdot)$ é a **função de momento** (ou **equação de estimação**) que relaciona $(W,\theta)$;
- $P$ é a **lei populacional de $W$** (a distribuição conjunta verdadeira),
  - $P=\mathcal{L}(Y,D,X)$, portanto $W\sim P$.

A ideia central é: **o parâmetro $\theta_0$ é definido (e/ou identificado) como o valor que zera um momento populacional**.



### OLS como projeção e como condição de momento

#### OLS como problema de projeção (amostra fixa)

No modelo linear (em amostra),

$$
\mathbf y=\mathbf X\beta+\mathbf u,
$$

com:

- $\mathbf y\in\mathbb R^n$,
- $\mathbf X\in\mathbb R^{n\times k}$,
- $\beta\in\mathbb R^k$.

O OLS escolhe $\hat\beta$ para minimizar a distância euclidiana entre $\mathbf y$ e o subespaço gerado pelas colunas de $\mathbf X$:

$$
\hat\beta=\arg\min_{\beta\in\mathbb R^k}\|\mathbf y-\mathbf X\beta\|^2.
$$

Defina o resíduo estimado:

$$
\hat{\mathbf u}=\mathbf y-\mathbf X\hat\beta.
$$

**Condição de otimalidade (equações normais):**

$$
\mathbf X'(\mathbf y-\mathbf X\hat\beta)=\mathbf 0
\quad\Longleftrightarrow\quad
\mathbf X'\hat{\mathbf u}=\mathbf 0.
$$

- Interpretação geométrica: **o resíduo $\hat{\mathbf u}$ é ortogonal ao espaço coluna de $\mathbf X$**.
- Importante: isso vale **sempre** (dado $\mathbf X$ com posto cheio). Não é hipótese; é consequência do problema de minimização.


#### De ortogonalidade a momento amostral

Escrevendo por observação (linha $i$):

$$
\hat u_i = y_i - x_i'\hat\beta,
$$

onde $x_i\in\mathbb R^k$ é o vetor de regressoras da observação $i$.

A condição de otimalidade pode ser escrita como um **momento amostral**:

$$
\frac{1}{n}\sum_{i=1}^n x_i\,(y_i-x_i'\hat\beta)=0.
$$

Defina a função de momento:

$$
\psi(W_i,\beta)=x_i\,(y_i-x_i'\beta).
$$

Então o OLS resolve:

$$
\boxed{\frac{1}{n}\sum_{i=1}^n \psi(W_i,\hat\beta)=0.}
$$

Isso **é** uma **condição de momento amostral**.


### E o momento populacional? (onde entra “exogeneidade”)

Agora entra a parte probabilística: existe uma lei populacional $P$ que gera $W$. Muitas vezes chamamos essa lei de **Processo Gerador de Dados** (*Data Generating Process*, DGP).

No modelo estrutural populacional:

$$
Y=X'\beta_0+\varepsilon,
$$

onde $\varepsilon$ é o erro **populacional**.

Uma condição típica de identificação (exogeneidade) é:

$$
E[\varepsilon\mid X]=0
\quad\Longrightarrow\quad
E[X\varepsilon]=0.
$$

Note a ponte:

$$
E[X(Y-X'\beta_0)] = E[X\varepsilon]=0.
$$

Ou seja, sob a hipótese $E[\varepsilon\mid X]=0$, o parâmetro estrutural $\beta_0$ satisfaz o **momento populacional**:

$$
\boxed{E_P[\psi(W,\beta_0)]=0.}
$$

E como $P$ é desconhecida, aproximamos a expectativa populacional por média amostral (análogo amostral):

$$
E_P[\psi(W,\beta)]\approx \frac{1}{n}\sum_{i=1}^n \psi(W_i,\beta),
$$

o que nos devolve exatamente o OLS estimado.


### Expectativa como funcional: por que aparece $E_P[g(W)]$?

Por que temos algo como $E_{P}[g(W)]$, para uma função qualquer $g$ que queiramos medir?

Repare que há dois ingredientes:

- a função $g$ que queremos medir;
- a distribuição $P$ que gera os dados $W$.

Fixando a distribuição de probabilidade $P$ que gera $W$. Definimos o operador:

$$
T_{P}: g \mapsto E_{P}[g(W)] = \int g(w)\, dP(w).
$$

Repare:
- **entrada**: uma função $g$;
- **saída**: um número real (a expectativa de $g(W)$).

Isso é exatamente um **funcional** (um funcional linear no espaço de funções integráveis sob $P$).

> A expectativa é um funcional: para uma distribuição $P$, o operador $T_P$ pega uma função $g$ e devolve um número real, que é a expectativa de $g(W)$ sob $P$. Fixando $g$, o número $E_P[g(W)]$ também pode ser visto como um funcional de $P$.

Fixando a função $g$, o mapeamento

$$
\Theta(P) := E_{P}[g(W)]
$$

é um **funcional escalar** da distribuição $P$ (":=" significa "definido como"). Pense em $\Theta(.)$ como uma função, só que a “entrada” não é um número $x$: a entrada é uma distribuição $P$. O “output” é o parâmetro/estimando que você quer.

* Se $k=1$, $\Theta(P)$ é um número real (parâmetro escalar);
* Se $k>1$, $\Theta(P)$ é um vetor em $\mathbb{R}^k$ (parâmetro vetorial).
* Pode ser uma matriz, ou até uma função.
* $\Theta(P)$ é o valor do parâmetro/estimando no mundo $P$.


Exemplos:

- média: $\Theta(P)=E_P[Y]$;
- variância: $\Theta(P)=E_P[(Y-E_P[Y])^2]$.

No OLS, o parâmetro $\beta$ tipicamente é definido implicitamente por um momento:

$$
m(P,\beta) := E_P[X(Y-X'\beta)] = 0.
$$

Aí o “funcional de interesse” é

$$
\beta(P) := \{\beta\in\mathbb{R}^p: m(P,\beta)=0\}.
$$

Sob condições usuais (ex.: $E_P[XX']$ inversível), isso vira explícito:

$$
\beta(P) = (E_P[XX'])^{-1} E_P[XY] \in \mathbb{R}^p.
$$

Aqui fica bem claro:
* $\mathbb{R}^p$ é o espaço onde o parâmetro vive (codomain).
* $\beta(P)$ é um vetor específico (o valor populacional “verdadeiro”) quando a distribuição é $P$.


> Quando eu fixo $g$, escolho qual característica quero medir (média, variância, um momento, um indicador). A partir daí, a expectativa vira uma função da distribuição $P$; logo, diferentes populações geram diferentes valores.


### Quando um momento “vira” uma condição de momento

Reforçando:

> Quando impomos que um momento seja zero (ou igual a algum valor), obtemos uma **condição de momento** que define/identifica parâmetros.

Em particular:

$$
\theta_0 \text{ é definido por } E_P[\psi(W,\theta_0)] = 0.
$$

Em outras palavras, o parâmetro (causal ou estrutural) $\theta_0$ é identificado se existe um **funcional** $\Theta: \mathcal P \to \mathbb R^d$ tal que:

$$
\theta_0=\Theta(P),
$$

onde $\Theta(P)$ é o valor que zera a condição de momento.

Exemplos de condições de momento:

- **Exogeneidade/OLS**:
  $$
  E_P[X\varepsilon] = E_P[X(Y-X'\beta_0)] = 0.
  $$

- **Instrumentos/IV**:
  $$
  E_P[Z\varepsilon] = E_P[Z(Y-X'\beta_0)] = 0,
  $$
  onde $Z$ é um vetor de instrumentos (e $Z$ é relevante para $X$).


### Mais conceitos

#### Funcional e momentos
- Momentos são expectativas do tipo $E_P[g(W)]$.
- Condições de momento definem parâmetros ao impor que um momento seja zero (ou igual a algum valor).

Para evitar confusão de notação, separe:

- **Momento populacional (como função de $\theta$):**
  $$
  M_P(\theta) := E_P[\psi(W,\theta)].
  $$

- **Parâmetro verdadeiro (funcional de $P$):**
  $$
  \theta_0=\Theta(P)\quad \text{tal que}\quad M_P(\theta_0)=0.
  $$


### Parâmetros vs Estimadores

- **Parâmetro ($\theta_0$):** valor verdadeiro (da população), fixo mas desconhecido; definido por condição de momento populacional.
  $$
  \theta_0=\Theta(P).
  $$

- **Estimador ($\hat\theta$):** regra de cálculo baseada na amostra; é uma variável aleatória (depende de $W_1,\dots,W_n$); definido por condição de momento amostral.

Se $\hat P_n$ é a distribuição empírica, então:

$$
\hat\theta=\Theta(\hat P_n),
\quad \text{isto é,}\quad
E_{\hat P_n}[\psi(W,\hat\theta)]=0
\ \Longleftrightarrow\
\frac{1}{n}\sum_{i=1}^n \psi(W_i,\hat\theta)=0.
$$



### Inferência Moderna

Para fazer inferência, queremos entender **como** $\hat\theta$ flutua em torno de $\theta_0$:

$$
\hat\theta-\theta_0=\Theta(\hat P_n)-\Theta(P).
$$

A pergunta conceitual central é:

> **Como pequenas mudanças na distribuição dos dados afetam o valor do parâmetro $\Theta(\hat{P})$?**

Essa pergunta é a porta de entrada para a **função de influência**.

### Mudança infinitesimal na lei populacional

Escolha um ponto específico $w$ no espaço amostral (por exemplo, $w=(y,d,x)$).
Considere a distribuição perturbada

$$
P_\varepsilon = (1-\varepsilon)P + \varepsilon\,\delta_w,
\qquad \varepsilon\in(0,1).
$$

Interpretação:

- Com probabilidade $(1-\varepsilon)$, o dado vem de $P$.
- Com probabilidade $\varepsilon$, o dado é **forçado** a ser exatamente $w$.
- $\delta_w$ é a **medida de Dirac**: coloca toda a massa no ponto $w$.

Essa é uma forma precisa de formalizar:

> **“E se eu der um peso infinitesimal a uma observação específica?”**


### Definição: Função de Influência (Influence Function - IF)

Se $\Theta$ for “suave” o suficiente, podemos olhar a derivada de $\Theta(P_\varepsilon)$ em $\varepsilon=0$.

A **função de influência** de $\Theta$ em $P$, avaliada no ponto $w$, é definida como:

$$
\boxed{
IF(w; \Theta, P)
\;:=\;
\left.\frac{d}{d\varepsilon}\Theta(P_\varepsilon)\right|_{\varepsilon=0}.
}
$$

Interpretação:

- $IF(w;\Theta,P)$ mede o **efeito marginal** (de primeira ordem) de “injetar” um peso infinitesimal em $w$.
- Observações com $|IF(w)|$ grande são, localmente, **muito influentes** para o parâmetro.

Em muitas referências, escreve-se $\varphi(w)$ em vez de $IF(w;\Theta,P)$:
$$
\varphi(w) \equiv IF(w;\Theta,P).
$$

**Exemplo 1: Média**

Se $\Theta(P)=E_P[Y]$, então sob $P_\varepsilon$:

$$
\Theta(P_\varepsilon)
= (1-\varepsilon)E_P[Y] + \varepsilon\,y.
$$

Logo,

$$
IF(w;\Theta,P)
=
\left.\frac{d}{d\varepsilon}\Theta(P_\varepsilon)\right|_{\varepsilon=0}
=
y - E_P[Y].
$$

Ou seja, para a média, a IF é simplesmente “desvio em relação à média”.


**Exemplo 2: OLS**

Considere ($W=(Y,X)$), com ($X\in\mathbb{R}^p$) (inclua intercepto em ($X$), se quiser). Defina o funcional OLS:

$$
\beta(P) := \arg\min_{\beta\in\mathbb{R}^p} E_P\left[(Y-X'\beta)^2\right]
\quad\Longleftrightarrow\quad
E_P[X(Y-X'\beta)]=0.
$$

Sob condições usuais (ex.: ($A:=E_P[XX']$) inversível), isso equivale a

$$
\beta(P) = A^{-1}c,\qquad
A:=E_P[XX'],\quad c:=E_P[XY].
$$

Agora faça a contaminação pontual no ponto ($w=(y,x)$):

$$
P_\varepsilon := (1-\varepsilon)P + \varepsilon \delta_{(y,x)}.
$$

***Passo 1: momentos sob ($P_\varepsilon$)***

$$
A_\varepsilon = E_{P_\varepsilon}[XX'] = (1-\varepsilon)A + \varepsilon xx'
$$

$$
c_\varepsilon = E_{P_\varepsilon}[XY] = (1-\varepsilon)c + \varepsilon xy
$$

Logo,

$$
\beta(P_\varepsilon)=\beta_\varepsilon = A_\varepsilon^{-1}c_\varepsilon.
$$

***Passo 2: derivada em ($\varepsilon=0$)***

$$
\beta'(P_\varepsilon)=(A^{-1})'c + A^{-1}c'.
$$


Use a regra (matricial) do produto:

$$
\frac{d}{d\varepsilon}(A_\varepsilon^{-1}c_\varepsilon) = (-A_\varepsilon^{-1}A_\varepsilon'A_\varepsilon^{-1}) c_\varepsilon + A_\varepsilon^{-1}c_\varepsilon'
$$

OBS: a derivada de $\frac{d}{d\varepsilon}(A_\varepsilon^{-1}c_\varepsilon)$ é entendida a partir da identidade $A(\varepsilon) (A(\varepsilon)^{-1}) = I$ (matriz identidade), e posterior derivação com relação a $\varepsilon$, a qual é igual a zero. Assim, é possível isolar a derivada de $A(\varepsilon)^{-1}$, chegando na fórmula acima.


Em ($\varepsilon=0$): ($A_0=A$), ($c_0=c$), ($\beta_0=\beta$), e

$$
A_\varepsilon'|_{0}=xx'-A,\qquad c_\varepsilon'|_{0}=xy-c.
$$

Então:
$$
\left.\frac{d\beta_\varepsilon}{d\varepsilon}\right|_{0}= -A^{-1}(xx'-A)A^{-1}c + A^{-1}(xy-c).
$$

Como ($A^{-1}c=\beta$), isso vira:
$$
\left.\frac{d\beta_\varepsilon}{d\varepsilon}\right|_{0}=-A^{-1}(xx'-A)\beta + A^{-1}(xy-c).
$$

E usando ($c=A\beta$):
$$
A^{-1}(xy-c) = A^{-1}(xy'-A\beta)
$$


Agora faça a álgebra dentro do $A^{-1}$:

$$
-(xx'-A)\beta + (xy- A\beta) = -xx'\beta + A\beta + xy - A\beta = xy - xx'\beta
$$

Logo, 

$$
\boxed{\frac{d\beta_{\varepsilon}}{d\varepsilon}\bigg|_{\varepsilon=0} = A^{-1}(xy - xx'\beta)}
$$

e, como $xy - xx'\beta = x(y-x'\beta)$, isso é equivalente a:

$$
\boxed{\frac{d\beta_{\varepsilon}}{d\varepsilon}\bigg|_{\varepsilon=0} = A^{-1}x(y-x'\beta)}
$$


### Resultado (Influence Function)

$$
\boxed{
IF\big((y,x);\beta,P\big)=A^{-1}x(y-x'\beta)
}
\quad\in\mathbb{R}^p.
$$

***Interpretação rápida***

* ($u:=y-x'\beta$) é o **resíduo populacional** no ponto contaminante.
* ($A^{-1}x$) é é o fator de sensibilidade direcional: como um choque no momento $x(y-x'\beta)$ se traduz em variação em $\beta$, dada a geometria populacional (via $E[XX']$).
* Então a influência é “**resíduo ($\times$) alavancagem**”, generalizando bem a intuição da média.


#### Como IF e ($\psi$) se relacionam

Para muitos M-estimadores (maximum/minimum) definidos por ($E[\psi(W,\theta_0)]=0$), vale:

$$
\boxed{ IF(w;\theta_0,P_0) = -\Big(E[\partial_\theta \psi(W,\theta_0)]\Big)^{-1}\psi(w,\theta_0)}
$$

Ou seja: **a IF é uma versão “re-escalada” da equação de estimação ($\psi$)** pelo inverso do Jacobiano.

* No OLS: ($\psi=x(y-x'\beta)$), ($E[\partial_\beta \psi]= -E[XX']$), então
$$
IF = (E[XX'])^{-1}x(y-x'\beta).
$$

Isso explica por que, em muita literatura, as pessoas chamam ($\psi$) de “score” e depois tratam a IF como “score eficiente” ou “escore ortogonal” — mas tecnicamente são **objetos diferentes**. 

**Intuição:**
- $\psi(w,\theta_0)$ é a contribuição daquela observação ($w$) para a equação de estimação.
- $E[\partial_\theta \psi(W,\theta_0)]$ é a sensibilidade do momento em relação a $\theta$.
- A IF re-escalona a contribuição de $w$ para refletir seu impacto na estimativa de $\theta$.
- Ou seja, temos uma mudança em $\theta$ que é o produto entre a contribuição de $w$ para o momento e a sensibilidade média na -1.

### Inferência Moderna

Suponha que $W_{1}, ..., W_{n}$ são i.i.d. de $P$. Se o parâmetro é um funcional $\theta_0=\Theta(P)$ e o estimador é $\hat\theta=\Theta(\hat P_n)$, então

$$
\hat\theta-\theta_0=\Theta(\hat P_n)-\Theta(P).
$$

A pergunta conceitual central é:

> **Como pequenas mudanças na distribuição dos dados afetam o valor do parâmetro $\Theta(P)$?**

Essa pergunta é a porta de entrada para a **função de influência**.


#### Linearização (expansão de von Mises)

Sob condições de regularidade (diferenciabilidade de $\Theta$ em $P$), existe uma função $IF(\cdot;\Theta,P)$ tal que

$$
\Theta(\hat P_n)-\Theta(P)
=
\int IF(w;\Theta,P)\,d(\hat P_n-P)(w)
+
o_p(n^{-1/2}).
$$

Como $\hat P_n=\frac{1}{n}\sum_{i=1}^n \delta_{W_i}$, isso implica

$$
\Theta(\hat P_n)-\Theta(P)
=
\frac{1}{n}\sum_{i=1}^n IF(W_i;\Theta,P)
-
E_{P}[IF(W;\Theta,P)]
+
o_p(n^{-1/2}).
$$

Sem perda de generalidade, podemos tomar a versão centrada da função de influência, isto é,

$$
E_{P}[IF(W;\Theta,P)]=0
$$

Logo,

$$
\boxed{
\hat\theta-\theta_0
=
\frac{1}{n}\sum_{i=1}^n IF(W_i;\Theta,P)
+
o_p(n^{-1/2}).
}
$$

Essa é a representação-chave: o erro de estimação é a média de uma variável aleatória (a IF) + um termo assintoticamente desprezível.

- $o_p(n^{-1/2})$ significa que o termo de resto é pequeno em comparação com $n^{-1/2}$, em probabilidade.
- A escala $n^{-1/2}$ é a taxa usual para estimadores regulares porque $\frac{1}{n}\sum IF(W_i; \Theta, P)$ é uma média amostral, e médias amostrais flutuam tipicamente na ordem de $n^{-1/2}$ (Teorema Central do Limite), levando à normalidade assintótica.

#### Alternativa intuitiva: Taylor na equação de estimação (M-estimadores)

Muitos estimadores em econometria podem ser escritos como soluções de uma equação de estimação:

$$
0=\frac{1}{n}\sum_{i=1}^n \psi(W_i,\hat\theta)
\;=:\;
g_n(\hat\theta),
\qquad
\text{com } E_P[\psi(W,\theta_0)]=0.
$$

Aqui, $g_n(\theta)$ é a versão amostral do momento populacional $g(\theta):=E_P[\psi(W,\theta)]$.

##### Taylor de primeira ordem em torno de $\theta_0$

Faça uma expansão de Taylor de $g_n(\hat\theta)$ ao redor de $\theta_0$:

$$
0=g_n(\hat\theta)
\approx
g_n(\theta_0)
+
\partial_\theta g_n(\theta_0)\,(\hat\theta-\theta_0).
$$

Isolando $(\hat\theta-\theta_0)$:

$$
\hat\theta-\theta_0
\approx
-\Big(\partial_\theta g_n(\theta_0)\Big)^{-1}\,g_n(\theta_0).
$$

Agora, note que

$$
g_n(\theta_0)=\frac{1}{n}\sum_{i=1}^n \psi(W_i,\theta_0),
\qquad
\partial_\theta g_n(\theta_0)=\frac{1}{n}\sum_{i=1}^n \partial_\theta\psi(W_i,\theta_0).
$$

Sob LLN, $\partial_\theta g_n(\theta_0)\to J$, onde

$$
J:=E_P[\partial_\theta\psi(W,\theta_0)].
$$

Logo, obtemos a representação linear

$$
\hat\theta-\theta_0
=
\frac{1}{n}\sum_{i=1}^n \Big(-J^{-1}\psi(W_i,\theta_0)\Big)
+
o_p(n^{-1/2}).
$$

Portanto, a função de influência associada a esse M-estimador é

$$
\boxed{
IF(w;\theta_0,P)= -J^{-1}\psi(w,\theta_0),
\qquad
J=E_P[\partial_\theta\psi(W,\theta_0)].
}
$$

##### Comentário (conexão com von Mises)

A expansão de von Mises faz a mesma coisa, mas “por trás”: ela lineariza o funcional $\Theta(P)$ quando trocamos $P$ por $\hat P_n$. No caso de M-estimadores, a Taylor na equação $g_n(\theta)=0$ produz explicitamente o mesmo termo líder (a IF), que é o objeto usado para obter SE e IC via CLT.

#### Normalidade assintótica (CLT) e variância assintótica

Da representação

$$
\hat\theta-\theta_0
=
\frac{1}{n}\sum_{i=1}^n IF(W_i;\Theta,P)
+
o_p(n^{-1/2}),
$$

multiplicando por $\sqrt{n}$ obtemos

$$
\sqrt{n}(\hat\theta-\theta_0)
=
\frac{1}{\sqrt{n}}\sum_{i=1}^n IF(W_i;\Theta,P)
+
\sqrt{n}\,o_p(n^{-1/2})
=
\frac{1}{\sqrt{n}}\sum_{i=1}^n IF(W_i;\Theta,P)
+
o_p(1).
$$

Defina $IF_i := IF(W_i;\Theta,P)$. Suponha $E_P[IF(W;\Theta,P)]=0$ e $Var_P(IF(W;\Theta,P))=\sigma^2<\infty$ (no caso escalar), e que $\{W_i\}$ são i.i.d.

***Variância da soma: cresce com $n$***

Como os $IF_i$ são independentes,

$$
Var\!\left(\sum_{i=1}^n IF_i\right)
=
\sum_{i=1}^n Var(IF_i)
=
n\sigma^2.
$$

Ou seja: a variância da soma cresce linearmente em $n$. Ao dividirmos por $\sqrt{n}$, esse crescimento é cancelado:

$$
Var\!\left(\frac{1}{\sqrt{n}}\sum_{i=1}^n IF_i\right)
=
\frac{1}{n}\,Var\!\left(\sum_{i=1}^n IF_i\right)
=
\frac{1}{n}\,n\sigma^2
=
\sigma^2.
$$

Então o “tamanho típico” do ruído fica estável, e é por isso que o CLT dá um limite não-degenerado.

Assim, pelo Teorema Central do Limite,

$$
\boxed{
\sqrt{n}(\hat\theta-\theta_0)
\Rightarrow
\mathcal{N}(0,\Sigma),
\qquad
\Sigma:=Var_P\!\big(IF(W;\Theta,P)\big).
}
$$

Em particular, para um parâmetro escalar ($\theta\in\mathbb{R}$),

$$
\Sigma = Var_P(IF(W)) = E_P[IF(W)^2],
$$

pois $E_P[IF(W)]=0$.

Para um parâmetro vetorial ($\theta\in\mathbb{R}^k$), $IF(W)\in\mathbb{R}^k$ e

$$
\Sigma = Var_P(IF(W)) = E_P\!\left[ IF(W)\,IF(W)'\right],
\quad\text{(matriz $k\times k$)}.
$$

Assim, o erro de estimação tem ordem típica $n^{-1/2}$ e a variância assintótica do estimador decresce com $n$:

$$
Var(\hat\theta)\approx \frac{1}{n}\Sigma.
$$

Repare que a variância do estimador não é constante: ela decresce com $n$ (taxa $1/n$). O que fica estável é a variância do erro **normalizado** $\sqrt{n}(\hat\theta-\theta_0)$.

### Construir SE e IC

#### Passo 1 — construir $\widehat{IF}_i$
A IF verdadeira depende de $P$ e de $\theta_0$, então usamos *plug-in*:

$$
\widehat{IF}_i := IF(W_i;\hat\theta,\hat P_n),
$$

isto é, substituímos quantidades populacionais por amostrais e $\theta_0$ por $\hat\theta$.

#### Passo 2 — estimar $Var(\hat\theta)$
Para $\theta\in\mathbb{R}^k$ (vetor), estimamos:

$$
\boxed{
\widehat{Var}(\hat\theta)
=
\frac{1}{n^2}\sum_{i=1}^n \widehat{IF}_i\,\widehat{IF}_i'
}
\qquad (k\times k).
$$

Para $\theta$ escalar:

$$
\boxed{
\widehat{Var}(\hat\theta)
=
\frac{1}{n^2}\sum_{i=1}^n \widehat{IF}_i^2
},
\qquad
se(\hat\theta)=\sqrt{\widehat{Var}(\hat\theta)}.
$$

Para o componente $j$ de um vetor $\hat\theta$:

$$
\boxed{
se(\hat\theta_j)=\sqrt{\left[\widehat{Var}(\hat\theta)\right]_{jj}}.
}
$$

### Passo 3 — IC (Wald) para o componente $j$
Para nível $1-\alpha$:

$$
\boxed{
IC_{1-\alpha}(\theta_j):
\quad
\hat\theta_j \pm z_{1-\alpha/2}\,se(\hat\theta_j).
}
$$

#### Caso comum: M-estimadores definidos por $\psi$ (just-identified)

Se $\hat\theta$ resolve

$$
\frac{1}{n}\sum_{i=1}^n \psi(W_i,\hat\theta)=0,
$$

então tipicamente a IF é

$$
IF(w;\theta,P)= -J^{-1}\psi(w,\theta),
\qquad
J:=E_P[\partial_\theta \psi(W,\theta)].
$$

Implementação plug-in:

$$
\hat J=\frac{1}{n}\sum_{i=1}^n \partial_\theta \psi(W_i,\hat\theta),
\qquad
\widehat{IF}_i = -\hat J^{-1}\psi(W_i,\hat\theta).
$$

E então:

$$
\widehat{Var}(\hat\theta)=\frac{1}{n^2}\sum_{i=1}^n \widehat{IF}_i\widehat{IF}_i'.
$$

### Exemplo: OLS (robusto a heterocedasticidade) via IF

No OLS:

$$
\widehat{IF}_i=\hat A^{-1}X_i\hat u_i,
\qquad
\hat A=\frac{1}{n}\sum_{i=1}^n X_iX_i',
\qquad
\hat u_i=Y_i-X_i'\hat\beta.
$$

Logo:

$$
\widehat{Var}(\hat\beta)
=
\frac{1}{n^2}\sum_{i=1}^n
(\hat A^{-1}X_i\hat u_i)(\hat A^{-1}X_i\hat u_i)'.
$$


### Extensão: dados em painel / erros clusterizados

Se há clusters $g=1,\dots,G$ (ex.: unidade $i$ em painel), agregue a IF por cluster:

$$
\hat S_g := \sum_{i\in g}\widehat{IF}_i.
$$

Então:

$$
\boxed{
\widehat{Var}_{cl}(\hat\theta)
=
\frac{1}{n^2}\sum_{g=1}^G \hat S_g\hat S_g'.
}
$$
